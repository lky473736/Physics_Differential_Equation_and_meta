```python
import os
import random
import time
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from thop import profile
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import re
from collections import Counter
from sklearn.preprocessing import LabelEncoder, StandardScaler
import glob as glob_module
import pickle
import scipy.io as scio
from sklearn.model_selection import KFold, train_test_split


def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True

def compute_mean_std(X):
    X = X.astype(np.float32)
    mean = X.mean(axis=(0, 2), keepdims=True)
    std  = X.std(axis=(0, 2), keepdims=True) + 1e-8
    return mean, std

def aug_jitter_scale(x_batch, jitter_sigma=0.03, scale_max=0.1):
    jitter = torch.randn_like(x_batch) * jitter_sigma
    x_batch = x_batch + jitter
    scale = 1.0 + (torch.rand(x_batch.shape[0], 1, 1, device=x_batch.device) - 0.5) * 2 * scale_max
    x_batch = x_batch * scale
    return x_batch

def aug_sensor_failure(x_batch):
    B, C, L = x_batch.shape
    x_aug = x_batch.clone()
    fail_channel = random.randint(0, C - 1)
    x_aug[:, fail_channel, :] = 0.0
    return x_aug

def aug_missing_segment(x_batch, missing_seg_len_range=(0.1, 0.25)):
    B, C, L = x_batch.shape
    x_aug = x_batch.clone()
    seg_len_min = int(L * missing_seg_len_range[0])
    seg_len_max = int(L * missing_seg_len_range[1])
    if seg_len_min >= seg_len_max:
         seg_len_max = seg_len_min + 1
    seg_len = random.randint(seg_len_min, seg_len_max)
    start = random.randint(0, L - seg_len)
    x_aug[:, :, start:start+seg_len] = 0.0
    return x_aug

def aug_outlier_spike(x_batch, outlier_magnitude=10, num_outliers_range=(1, 5)):
    B, C, L = x_batch.shape
    x_aug = x_batch.clone()
    device = x_batch.device

    num_outliers = random.randint(num_outliers_range[0], num_outliers_range[1])
    if num_outliers == 0:
        return x_aug

    b_idx = torch.randint(0, B, (num_outliers,), device=device)
    c_idx = torch.randint(0, C, (num_outliers,), device=device)
    l_idx = torch.randint(0, L, (num_outliers,), device=device)

    magnitudes = torch.randn(num_outliers, device=device) * outlier_magnitude

    flat_indices = b_idx * (C * L) + c_idx * L + l_idx
    x_aug.view(-1).index_add_(0, flat_indices, magnitudes)

    return x_aug

def apply_test_noise(x_batch, noise_type, noise_level):
    x_noisy = x_batch.clone()
    B, C, L = x_batch.shape
    device = x_batch.device

    if noise_type == 'jitter':
        noise = torch.randn_like(x_batch) * noise_level
        x_noisy = x_batch + noise

    elif noise_type == 'time_shift':
        max_shift = int(L * noise_level)
        if max_shift == 0 and noise_level > 0:
             max_shift = 1

        shifts = torch.randint(-max_shift, max_shift + 1, (B,), device=device)

        src_indices = torch.arange(L, device=device).unsqueeze(0).expand(B, -1)
        dest_indices = src_indices + shifts.unsqueeze(1)

        valid_mask = (dest_indices >= 0) & (dest_indices < L)

        batch_idx = torch.arange(B, device=device).view(-1, 1, 1).expand(-1, C, L)
        channel_idx = torch.arange(C, device=device).view(1, -1, 1).expand(B, -1, L)

        x_shifted = torch.zeros_like(x_batch)

        dest_time_idx = dest_indices.unsqueeze(1).expand(-1, C, -1)
        valid_mask_expanded = valid_mask.unsqueeze(1).expand(-1, C, -1)

        valid_dest_indices = dest_time_idx[valid_mask_expanded]
        valid_batch_indices = batch_idx[valid_mask_expanded]
        valid_channel_indices = channel_idx[valid_mask_expanded]
        valid_src_values = x_batch[valid_mask_expanded]

        x_shifted[valid_batch_indices, valid_channel_indices, valid_dest_indices] = valid_src_values
        x_noisy = x_shifted

    elif noise_type == 'sensor_failure':
        n_channels = int(np.ceil(C * noise_level))
        if n_channels == 0: n_channels = 1
        fail_channels = np.random.choice(C, n_channels, replace=False)
        x_noisy[:, fail_channels, :] = 0.0

    elif noise_type == 'outlier_spike':
        num_spikes = max(1, int(noise_level * 100))
        total_spikes = B * num_spikes

        b_idx = torch.arange(B, device=device).repeat_interleave(num_spikes)
        c_idx = torch.randint(0, C, (total_spikes,), device=device)
        l_idx = torch.randint(0, L, (total_spikes,), device=device)
        magnitudes = (torch.randn(total_spikes, device=device) * 5)

        flat_indices = b_idx * (C * L) + c_idx * L + l_idx
        x_noisy.view(-1).index_add_(0, flat_indices, magnitudes)

    elif noise_type == 'missing_segment':
        seg_len = max(1, int(L * noise_level))
        start = random.randint(0, L - seg_len)
        x_noisy[:, :, start:start+seg_len] = 0.0

    return x_noisy

def compute_encoder_stats(encoder, input_shape=(1,9,128)):
    device = next(encoder.parameters()).device
    macs, params = profile(encoder, inputs=(torch.randn(*input_shape).to(device),), verbose=False)
    return params/1e6, macs/1e6

def measure_meta_inference_time(model, n_way, sz=(1,9,128), dev='cuda', runs=100, warm=10):
    model.eval()
    x_q = torch.randn(sz, device=dev)
    x_s = torch.randn((n_way*5,) + sz[1:], device=dev)
    y_s = torch.tensor(np.repeat(range(n_way), 5), device=dev)

    with torch.no_grad():
        for _ in range(warm):
            _ = model(x_q, x_s, y_s, n_way)
    torch.cuda.synchronize() if dev=='cuda' else None
    t0 = time.time()
    with torch.no_grad():
        for _ in range(runs):
            _ = model(x_q, x_s, y_s, n_way)
    torch.cuda.synchronize() if dev=='cuda' else None
    return (time.time()-t0)/runs*1000

def compute_supervised_stats(model: nn.Module, input_shape=(9, 128)):
    device = next(model.parameters()).device
    dummy_input = torch.randn(1, *input_shape).to(device)
    model.eval()

    if not hasattr(model, 'classifier') or not hasattr(model, 'extract_features'):
        raise ValueError("Model must have 'classifier' attribute and 'extract_features' method.")

    classifier_original = model.classifier
    model.classifier = nn.Identity()
    macs_backbone, params_backbone = profile(model, inputs=(dummy_input,), verbose=False)
    model.classifier = classifier_original

    with torch.no_grad():
        features = model.extract_features(dummy_input)
        if isinstance(features, tuple):
            features = features[0]

    macs_cls, params_cls = profile(model.classifier, inputs=(features,), verbose=False)

    total_macs_m = (macs_backbone + macs_cls) / 1e6
    total_params_m = (params_backbone + params_cls) / 1e6
    return total_params_m, total_macs_m

def measure_supervised_inference_time(model: nn.Module, input_size=(1, 9, 128), device='cuda', n_runs=100, warmup=10):
    model.eval()
    inputs = torch.randn(input_size).to(device)

    with torch.inference_mode():
        for _ in range(warmup):
            _ = model(inputs)

    if device == 'cuda':
        torch.cuda.synchronize()

    start = time.time()
    with torch.inference_mode():
        for _ in range(n_runs):
            _ = model(inputs)

    if device == 'cuda':
        torch.cuda.synchronize()

    end = time.time()
    return (end - start) / n_runs * 1000

def plot_confusion_matrix(cm, labels):
    cmn = cm.astype('float')/cm.sum(axis=1)[:,None]
    df = pd.DataFrame(cmn, index=labels, columns=labels)
    plt.figure(figsize=(8,7))
    sns.heatmap(df, annot=True, fmt='.2f', cmap='Blues', cbar=True)
    plt.xlabel('Predicted', fontsize=12)
    plt.ylabel('True', fontsize=12)
    plt.title('Confusion Matrix (Normalized)', fontsize=14)
    plt.tight_layout()
    plt.show()

def plot_learning_curves(train_losses, train_accs, val_accs, title_suffix=""):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    ax1.plot(train_losses, linewidth=2, label='Train Loss')
    ax1.set_xlabel('Episode', fontsize=12)
    ax1.set_ylabel('Loss', fontsize=12)
    ax1.set_title(f'Training Loss {title_suffix}', fontsize=14)
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    ax2.plot(train_accs, linewidth=2, color='orange', label='Train Acc')
    ax2.plot(val_accs, linewidth=2, color='green', label='Val Acc')
    ax2.set_xlabel('Episode', fontsize=12)
    ax2.set_ylabel('Accuracy', fontsize=12)
    ax2.set_title(f'Training & Validation Accuracy {title_suffix}', fontsize=14)
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    plt.tight_layout()

def plot_kshot_comparison(shot_results):
    plt.figure(figsize=(10, 6))
    plt.plot([r['K'] for r in shot_results], [r['Accuracy'] for r in shot_results],
             marker='o', linewidth=2, markersize=8, label='Accuracy')
    plt.plot([r['K'] for r in shot_results], [r['F1'] for r in shot_results],
             marker='s', linewidth=2, markersize=8, label='F1-Score')
    plt.xlabel('K-Shot', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.title('Meta-Learning Performance vs K-Shot', fontsize=14)
    plt.legend(fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_user_performance(per_user_df):
    plt.figure(figsize=(12, 6))
    x = range(len(per_user_df))
    plt.bar(x, per_user_df['Accuracy'], alpha=0.7, label='Accuracy')
    plt.axhline(y=per_user_df['Accuracy'].mean(), color='r', linestyle='--',
                label=f'Mean: {per_user_df["Accuracy"].mean():.3f}')
    plt.xlabel('Test User', fontsize=12)
    plt.ylabel('Accuracy', fontsize=12)
    plt.title('Per-User 5-Shot Accuracy', fontsize=14)
    plt.xticks(x, per_user_df['User'])
    plt.legend()
    plt.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.show()

class HARDataset(Dataset):
    def __init__(self, X, y, norm=True):
        self.X = X.astype(np.float32)
        self.y = y.astype(np.int64)
        if norm:
            self.mean = self.X.mean(axis=(0,2), keepdims=True)
            self.std = self.X.std(axis=(0,2), keepdims=True) + 1e-8
            self.X = (self.X - self.mean)/self.std
        else:
            self.mean, self.std = None, None

    def set_stats(self, mean, std):
        self.mean, self.std = mean, std
        self.X = (self.X - self.mean)/self.std

    def __len__(self):
        return len(self.y)

    def __getitem__(self, i):
        return torch.from_numpy(self.X[i]), int(self.y[i])

def read_txt_matrix(path):
    with open(path, 'r') as f:
        lines = f.readlines()
    lines = [list(map(float, line.strip().split())) for line in lines]
    return np.array(lines)

def slide_window(a, w, s):
    a = np.asarray(a)
    if a.shape[0] < w:
        return np.empty((0, w, a.shape[1]), dtype=a.dtype)
    sh = (a.shape[0] - w + 1, w, a.shape[1])
    st = (a.strides[0], a.strides[0], a.strides[1])
    A = np.lib.stride_tricks.as_strided(a, shape=sh, strides=st)
    return A[::s]

def load_opportunity_local_data(data_path="/content/drive/MyDrive/OpportunityUCIDataset/dataset", window_size=30, step=15):
    print("Loading Opportunity dataset from local files...")
    target_activities = {
        406516: 'Open Door 1', 406517: 'Open Door 2', 404516: 'Close Door 1', 404517: 'Close Door 2',
        406520: 'Open Fridge', 404520: 'Close Fridge', 406505: 'Open Dishwasher', 404505: 'Close Dishwasher',
        406519: 'Open Drawer 1', 404519: 'Close Drawer 1', 406511: 'Open Drawer 2', 404511: 'Close Drawer 2',
        406508: 'Open Drawer 3', 404508: 'Close Drawer 3', 408512: 'Clean Table', 407521: 'Drink from Cup',
        405506: 'Toggle Switch'
    }
    try:
        dat_files = glob_module.glob(os.path.join(data_path, "**", "*.dat"), recursive=True)
        if not dat_files:
            print(f"No .dat files found in {data_path}")
            return None, None, None, None
        print(f"Found {len(dat_files)} .dat files")

        all_data = []
        all_labels = []
        all_subjects_list = []

        for file_path in dat_files:
            print(f"  - Processing {os.path.basename(file_path)}...")
            match = re.search(r'S(\d+)-', os.path.basename(file_path))
            if not match:
                print(f"    Skipping {file_path}, no subject ID found in filename.")
                continue
            subject_id = int(match.group(1))

            try:
                data = pd.read_csv(file_path, sep=r'\s+', header=None, na_values='NaN')
                if data.empty or data.shape[1] < 250:
                    continue

                data = data.ffill().bfill()

                imu_cols = [38, 39, 40, 41, 42, 43, 51, 52, 53, 54, 55, 56, 64, 65, 66, 67, 68, 69, 77, 78, 79, 80, 81, 82, 90, 91, 92, 93, 94, 95]
                if data.shape[1] <= max(imu_cols):
                    print(f"    Skipping {file_path}, not enough columns.")
                    continue

                features = data.iloc[:, imu_cols].values
                labels = data.iloc[:, 249].values

                valid_mask = (np.isfinite(features).all(axis=1) & np.isfinite(labels) & ~np.isnan(labels))
                features = features[valid_mask]
                labels = labels[valid_mask]

                target_mask = np.isin(labels, list(target_activities.keys()))
                if not np.any(target_mask):
                    continue

                features = features[target_mask]
                labels = labels[target_mask]

                all_data.append(features)
                all_labels.extend(labels)
                all_subjects_list.extend([subject_id] * len(labels))

            except Exception as e:
                print(f"    Error processing {file_path}: {e}")
                continue

        if not all_data:
            print("No valid data loaded with target activities")
            return None, None, None, None

        X_combined = np.vstack(all_data)
        y_combined = np.array(all_labels)
        sub_combined = np.array(all_subjects_list)

        print(f"Combined data shape: {X_combined.shape}")
        print(f"Found activities: {np.unique(y_combined)}")

        unique_labels = np.unique(y_combined)
        label_mapping = {label: idx for idx, label in enumerate(unique_labels)}
        activity_labels = [target_activities.get(label, f"Class_{label}") for label in unique_labels]
        y_encoded = np.array([label_mapping[label] for label in y_combined])

        print(f"Label mapping: {label_mapping}")
        print(f"Activity labels: {activity_labels}")

        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X_combined)

        if np.any(np.isnan(X_scaled)) or np.any(np.isinf(X_scaled)):
            print("Cleaning NaN/inf values after scaling...")
            X_scaled = np.nan_to_num(X_scaled, nan=0.0, posinf=1.0, neginf=-1.0)

        all_windows = []
        all_window_labels = []
        all_window_subjects = []

        start = 0
        while start + window_size <= len(X_scaled):
            window_data = X_scaled[start:start + window_size, :]
            window_label = y_encoded[start + window_size - 1]
            window_subject = sub_combined[start + window_size - 1]

            if np.isfinite(window_data).all():
                all_windows.append(window_data)
                all_window_labels.append(window_label)
                all_window_subjects.append(window_subject)
            start += step

        if not all_windows:
            print("No valid windows created")
            return None, None, None, None

        X_sequences = np.array(all_windows, dtype=np.float32)
        X_sequences = np.transpose(X_sequences, (0, 2, 1))
        y_sequences = np.array(all_window_labels)
        sub_sequences = np.array(all_window_subjects)

        print(f"Final data validation:")
        print(f"Opportunity - X_sequences shape: {X_sequences.shape}")
        print(f"Opportunity - y_sequences shape: {y_sequences.shape}")
        print(f"Opportunity - sub_sequences shape: {sub_sequences.shape}")
        print(f"Number of classes: {len(activity_labels)}")
        print(f"Class distribution: {np.bincount(y_sequences)}")

        print("\n" + "=" * 50)
        print("Per-Subject Class Distribution (Window Count) - OPPORTUNITY")
        print("=" * 50)

        all_subjects_u = np.unique(sub_sequences)
        all_labels_idx = sorted(list(np.unique(y_sequences)))

        dist_data = []

        for sub_id in all_subjects_u:
            sub_mask = (sub_sequences == sub_id)
            sub_labels = y_sequences[sub_mask]
            class_counts = Counter(sub_labels)

            row = {'Subject': f"S{sub_id}"}
            total = 0
            for label_idx in all_labels_idx:
                count = class_counts.get(label_idx, 0)
                row[activity_labels[label_idx]] = count
                total += count
            row['Total'] = total
            dist_data.append(row)

        dist_df = pd.DataFrame(dist_data)
        dist_df = dist_df.set_index('Subject')

        col_order = [activity_labels[i] for i in all_labels_idx] + ['Total']
        dist_df = dist_df[col_order]

        display(dist_df)
        print("=" * 50 + "\n")

        return X_sequences, y_sequences, sub_sequences, activity_labels

    except Exception as e:
        print(f"Error loading Opportunity dataset: {e}")
        return None, None, None, None


def load_mobiact_sensor_file(path):
    data = []
    try:
        with open(path, 'r') as f:
            in_data_section = False
            for line in f:
                if line.startswith('@DATA'):
                    in_data_section = True
                    continue
                if not in_data_section:
                    continue

                parts = line.strip().split(',')
                if len(parts) == 4:
                    try:
                        data.append([float(p) for p in parts])
                    except ValueError:
                        continue
    except Exception as e:
        print(f"Error reading {path}: {e}")
        return None

    if not data:
        return None
    return np.array(data, dtype=np.float32)

def load_mobiact_data(data_path, w_s=100, stride=50):
    print(f"Loading MobiAct data with window={w_s}, stride={stride}")

    ADL_CODES = ['STD', 'WAL', 'JOG', 'JUM', 'STU', 'STN', 'SCH', 'CSI', 'CSO']
    FALL_CODES = ['FOL', 'FKL', 'BSC', 'SDL']
    ALL_ACTIVITIES = ADL_CODES + FALL_CODES

    label_encoder = LabelEncoder()
    label_encoder.fit(ALL_ACTIVITIES)
    activity_names = list(label_encoder.classes_)

    subject_dirs = sorted(glob_module.glob(os.path.join(data_path, 'sub*')))
    subjects_to_use = []

    for sub_dir in subject_dirs:
        if os.path.isdir(os.path.join(sub_dir, 'ADL')) and os.path.isdir(os.path.join(sub_dir, 'FALLS')):
            subjects_to_use.append(os.path.basename(sub_dir))

    print(f"Found {len(subjects_to_use)} subjects with both ADL and FALLS: {subjects_to_use}")

    all_data_for_scaler = []
    all_windows, all_labels, all_subjects = [], [], []

    temp_data_store = []

    for sub_name in subjects_to_use:
        subject_id_match = re.search(r'(\d+)', sub_name)
        if not subject_id_match:
            continue
        subject_id = int(subject_id_match.group(0))

        for act_type_folder in ['ADL', 'FALLS']:
            act_folders_path = os.path.join(data_path, sub_name, act_type_folder, '*')

            for act_folder_path in glob_module.glob(act_folders_path):
                act_code = os.path.basename(act_folder_path)
                if act_code not in ALL_ACTIVITIES:
                    continue

                label = label_encoder.transform([act_code])[0]

                trial_files = glob_module.glob(os.path.join(act_folder_path, f'{act_code}_acc_{subject_id}_*.txt'))
                trials = sorted(list(set([f.split('_')[-1].split('.')[0] for f in trial_files])))

                for trial in trials:
                    print(f"    - Processing MobiAct: Sub {subject_id} / {act_code} / Trial {trial}")
                    acc_file = os.path.join(act_folder_path, f'{act_code}_acc_{subject_id}_{trial}.txt')
                    gyro_file = os.path.join(act_folder_path, f'{act_code}_gyro_{subject_id}_{trial}.txt')
                    ori_file = os.path.join(act_folder_path, f'{act_code}_ori_{subject_id}_{trial}.txt')

                    if not all(os.path.exists(f) for f in [acc_file, gyro_file, ori_file]):
                        continue

                    data_acc = load_mobiact_sensor_file(acc_file)
                    data_gyro = load_mobiact_sensor_file(gyro_file)
                    data_ori = load_mobiact_sensor_file(ori_file)

                    if any(d is None for d in [data_acc, data_gyro, data_ori]):
                        continue

                    L = min(len(data_acc), len(data_gyro), len(data_ori))

                    if L < w_s:
                        continue

                    combined_data = np.hstack((data_acc[:L, 1:], data_gyro[:L, 1:], data_ori[:L, 1:]))
                    combined_data = combined_data.astype(np.float32)

                    combined_data = pd.DataFrame(combined_data).ffill().bfill().values

                    if np.isnan(combined_data).any():
                        continue

                    all_data_for_scaler.append(combined_data)
                    temp_data_store.append({
                        'data': combined_data,
                        'label': label,
                        'subject': subject_id
                    })

    if not all_data_for_scaler:
        print("No valid data found to fit scaler.")
        return None, None, None, None

    print("Fitting StandardScaler...")
    scaler = StandardScaler().fit(np.vstack(all_data_for_scaler))

    print("Applying scaling and windowing...")
    for item in temp_data_store:
        scaled_data = scaler.transform(item['data'])
        scaled_data = np.nan_to_num(scaled_data)

        windows = slide_window(scaled_data, w_s, stride)

        if windows.shape[0] > 0:
            all_windows.append(windows)
            all_labels.extend([item['label']] * windows.shape[0])
            all_subjects.extend([item['subject']] * windows.shape[0])

    if not all_windows:
        print("No valid windows created after processing.")
        return None, None, None, None

    X_windowed = np.vstack(all_windows).astype(np.float32)
    X_windowed = np.transpose(X_windowed, (0, 2, 1))
    y_encoded = np.array(all_labels, dtype=int)
    sub_array = np.array(all_subjects, dtype=int)

    print(f"MobiAct Loaded: X={X_windowed.shape}, y={y_encoded.shape}, sub={sub_array.shape}")

    print("\n" + "=" * 50)
    print("Per-Subject Class Distribution (Window Count) - MobiAct")
    print("=" * 50)

    all_subjects_u = np.unique(sub_array)
    all_labels_idx = sorted(list(np.unique(y_encoded)))

    dist_data = []

    for sub_id in all_subjects_u:
        sub_mask = (sub_array == sub_id)
        sub_labels = y_encoded[sub_mask]
        class_counts = Counter(sub_labels)

        row = {'Subject': f"S{sub_id}"}
        total = 0
        for label_idx in all_labels_idx:
            count = class_counts.get(label_idx, 0)
            row[activity_names[label_idx]] = count
            total += count
        row['Total'] = total
        dist_data.append(row)

    dist_df = pd.DataFrame(dist_data)
    dist_df = dist_df.set_index('Subject')

    col_order = [activity_names[i] for i in all_labels_idx] + ['Total']
    dist_df = dist_df[col_order]

    display(dist_df)
    print("=" * 50 + "\n")

    return X_windowed, y_encoded, sub_array, activity_names


def load_dsads_data(data_path, w_s=25, stride=12):
    print(f"Loading DSADS data with window={w_s}, stride={stride}")
    all_windows = []
    all_labels = []
    all_subjects = []

    activities = {
        'a01': 'sitting', 'a02': 'standing', 'a03': 'lying on back',
        'a04': 'lying on right side', 'a05': 'ascending stairs',
        'a06': 'descending stairs', 'a07': 'standing in an elevator still',
        'a08': 'moving around in an elevator', 'a09': 'walking in a parking lot',
        'a10': 'walking on a treadmill with a speed of 4 kmh',
        'a11': 'walking in flat and 15 deg inclined positions',
        'a12': 'running on a treadmill with a speed of 8 kmh',
        'a13': 'exercising on a stepper', 'a14': 'exercising on a cross trainer',
        'a15': 'cycling on an exercise bike in horizontal positions',
        'a16': 'cycling on an exercise bike in vertical positions',
        'a17': 'rowing', 'a18': 'jumping', 'a19': 'playing basketball'
    }
    activity_codes = sorted(activities.keys())

    label_encoder = LabelEncoder()
    label_encoder.fit(activity_codes)

    persons = ['p' + str(i) for i in range(1, 9)]

    for person_str in persons:
        subject_id = int(person_str[1:])

        for activity_str in activity_codes:
            activity_label = label_encoder.transform([activity_str])[0]
            pattern = os.path.join(data_path, activity_str, person_str, 's*.txt')
            segment_files = sorted(glob_module.glob(pattern))

            if not segment_files:
                continue

            for f in segment_files[:11]:
                print(f"    - Processing DSADS: {person_str} / {activity_str} / {os.path.basename(f)}")
                try:
                    segment_data = np.loadtxt(f, delimiter=',')
                    if segment_data.shape[0] < w_s:
                        continue

                    segment_data = pd.DataFrame(segment_data).ffill().bfill().values

                    start = 0
                    while start + w_s <= segment_data.shape[0]:
                        window_data = segment_data[start : start + w_s, :]
                        all_windows.append(window_data)
                        all_labels.append(activity_label)
                        all_subjects.append(subject_id)
                        start += stride

                except Exception as e:
                    print(f"Error loading or windowing {f}: {e}")

    if not all_windows:
        print("No valid windows created from DSADS data.")
        return None, None, None, None

    X_windowed = np.array(all_windows, dtype=np.float32)
    y_encoded = np.array(all_labels, dtype=int)
    subjects = np.array(all_subjects, dtype=int)

    scaler = StandardScaler()
    X_windowed_shape = X_windowed.shape
    X_windowed_flat = X_windowed.reshape(-1, X_windowed_shape[-1])
    X_scaled_flat = scaler.fit_transform(X_windowed_flat)
    X_windowed = X_scaled_flat.reshape(X_windowed_shape)

    X_windowed = np.transpose(X_windowed, (0, 2, 1))

    activity_names_sorted = [activities[code] for code in label_encoder.classes_]

    print("\n" + "=" * 50)
    print("Per-Subject Class Distribution (Window Count) - DSADS")
    print("=" * 50)

    all_subjects_u = np.unique(subjects)
    all_labels_idx = sorted(list(np.unique(y_encoded)))

    dist_data = []

    for sub_id in all_subjects_u:
        sub_mask = (subjects == sub_id)
        sub_labels = y_encoded[sub_mask]
        class_counts = Counter(sub_labels)

        row = {'Subject': f"S{sub_id}"}
        total = 0
        for label_idx in all_labels_idx:
            count = class_counts.get(label_idx, 0)
            row[activities[activity_codes[label_idx]]] = count
            total += count
        row['Total'] = total
        dist_data.append(row)

    dist_df = pd.DataFrame(dist_data)
    dist_df = dist_df.set_index('Subject')

    col_order = [activities[activity_codes[i]] for i in all_labels_idx] + ['Total']
    dist_df = dist_df[col_order]

    display(dist_df)
    print("=" * 50 + "\n")

    return X_windowed, y_encoded, subjects, activity_names_sorted

def load_uschad_data(path, w_s=100, stride=50):
    ACTIVITIES = {
        0: 'Walking Forward', 1: 'Walking Left', 2: 'Walking Right', 3: 'Walking Upstairs',
        4: 'Walking Downstairs', 5: 'Running Forward', 6: 'Jumping Up', 7: 'Sitting',
        8: 'Standing', 9: 'Sleeping', 10: 'Elevator Up', 11: 'Elevator Down'
    }

    result = []
    subject_ids = []
    labels = []

    if not os.path.exists(path):
        print(f"Path does not exist: {path}")
        return None, None, None, None

    subject_list = os.listdir(path)
    subject_list = [subject for subject in subject_list if subject.find('Subject') != -1]
    subject_list = sorted(subject_list)

    if not subject_list:
        print(f"No 'Subject' directories found in: {path}")
        return None, None, None, None

    print(f"Loading USC-HAD: Found {len(subject_list)} subjects. (w_s={w_s}, stride={stride})")

    original_cwd = os.getcwd()
    all_sensor_data = []

    try:
        os.chdir(path)

        for i, subject in enumerate(subject_list):
            if not os.path.isdir(subject):
                continue

            mat_list = os.listdir(subject)
            os.chdir(subject)

            for mat in mat_list:
                if not mat.endswith('.mat'):
                    continue

                content = scio.loadmat(mat)['sensor_readings']
                content = pd.DataFrame(content).ffill().bfill().values
                all_sensor_data.append(content)

            os.chdir('../')

        if not all_sensor_data:
            print("No sensor data found.")
            os.chdir(original_cwd)
            return None, None, None, None

        scaler = StandardScaler()
        scaler.fit(np.vstack(all_sensor_data))

        print("Scaler fitted. Loading, scaling, and windowing data...")
        os.chdir(path)

        for i, subject in enumerate(subject_list):
            if not os.path.isdir(subject):
                continue

            subject_num_match = re.findall(r'(\d+)', subject)
            if not subject_num_match:
                continue
            subject_num = int(subject_num_match[0])

            mat_list = os.listdir(subject)
            os.chdir(subject)

            for mat in mat_list:
                if not mat.endswith('.mat'):
                    continue

                print(f"    - Processing USC-HAD: {subject} / {mat}")

                activity_num_match = re.findall(r'a(\d+)', mat)
                if not activity_num_match:
                    continue

                category = int(activity_num_match[0]) - 1
                if category not in ACTIVITIES:
                    continue

                content = scio.loadmat(mat)['sensor_readings']
                content = pd.DataFrame(content).ffill().bfill().values

                if np.isnan(content).any():
                    continue

                content_scaled = scaler.transform(content)
                content_scaled = np.nan_to_num(content_scaled)

                x = slide_window(content_scaled, w_s, stride)

                result.extend(x)
                subject_ids.extend([subject_num] * len(x))
                labels.extend([category] * len(x))

            os.chdir('../')

    except Exception as e:
        print(f"Error loading data: {e}")
        os.chdir(original_cwd)
        return None, None, None, None

    os.chdir(original_cwd)

    if not result:
        print("No data loaded from .mat files.")
        return None, None, None, None

    data = np.array(result, dtype=np.float32)
    data = np.transpose(data, (0, 2, 1))

    activity_labels = [ACTIVITIES[i] for i in range(len(ACTIVITIES))]

    y_full = np.array(labels, dtype=int)
    sub_full = np.array(subject_ids, dtype=int)

    print("\n" + "=" * 50)
    print("Per-Subject Class Distribution (Window Count) - USC-HAD")
    print("=" * 50)

    all_subjects = np.unique(sub_full)
    all_labels_idx = sorted(list(ACTIVITIES.keys()))

    dist_data = []

    for sub_id in all_subjects:
        sub_mask = (sub_full == sub_id)
        sub_labels = y_full[sub_mask]
        class_counts = Counter(sub_labels)

        row = {'Subject': f"S{sub_id}"}
        total = 0
        for label_idx in all_labels_idx:
            count = class_counts.get(label_idx, 0)
            row[ACTIVITIES[label_idx]] = count
            total += count
        row['Total'] = total
        dist_data.append(row)

    dist_df = pd.DataFrame(dist_data)
    dist_df = dist_df.set_index('Subject')

    col_order = [ACTIVITIES[i] for i in all_labels_idx] + ['Total']
    dist_df = dist_df[col_order]

    display(dist_df)
    print("=" * 50 + "\n")

    return data, y_full, sub_full, activity_labels

def load_mhealth_data(dataset_dir="MHEALTH", **kwargs):
    subject_files = sorted([
        os.path.join(dataset_dir, f)
        for f in os.listdir(dataset_dir)
        if f.startswith("mHealth_subject") and f.endswith(".log")
    ])
    if not subject_files:
        print(f"No mHealth log files found in directory: {dataset_dir}")
        return None, None, None, None

    all_windows = []
    all_labels = []
    all_subjects = []
    window_size = 50
    step = 25

    print(f"Loading MHEALTH data with window={window_size}, stride={step}")

    all_features_list = []
    for file_path in subject_files:
        print(f"  - Pre-scanning MHEALTH: {os.path.basename(file_path)}")
        df = pd.read_csv(file_path, sep=r'\s+', header=None, engine='python', dtype=np.float32)
        df = df.ffill().bfill()
        if df.shape[1] < 24:
            continue
        labels = df.iloc[:, 23].values.astype(int)
        imu_cols = [0,1,2] + list(range(5, 23))
        features = df.iloc[:, imu_cols].values
        valid_indices = np.where(labels != 0)[0]
        if len(valid_indices) > 0:
            all_features_list.append(features[valid_indices, :])

    if not all_features_list:
        print("No valid features found in mHealth.")
        return None, None, None, None

    scaler = StandardScaler()
    scaler.fit(np.vstack(all_features_list))
    print("mHealth scaler fitted.")


    for file_path in subject_files:
        print(f"  - Processing MHEALTH: {os.path.basename(file_path)}")
        subject_id_match = re.search(r'subject(\d+)', os.path.basename(file_path))
        if not subject_id_match:
            continue
        subject_id = int(subject_id_match.group(1))

        df = pd.read_csv(file_path, sep=r'\s+', header=None, engine='python', dtype=np.float32)
        df = df.ffill().bfill()
        if df.shape[1] < 24:
            continue

        labels = df.iloc[:, 23].values.astype(int)
        imu_cols = [0,1,2] + list(range(5, 23))
        features = df.iloc[:, imu_cols].values

        valid_indices = np.where(labels != 0)[0]
        if len(valid_indices) == 0:
            continue

        features = features[valid_indices, :]
        labels = labels[valid_indices]

        if len(features) < window_size:
            continue

        features_scaled = scaler.transform(features)
        features_scaled = np.nan_to_num(features_scaled)

        start = 0
        while start + window_size <= len(features_scaled):
            window_data = features_scaled[start : start + window_size, :]
            window_labels_raw = labels[start : start + window_size]
            most_common_label = Counter(window_labels_raw).most_common(1)[0][0]

            all_windows.append(window_data)
            all_labels.append(most_common_label)
            all_subjects.append(subject_id)

            start += step

    if not all_windows:
        print("No valid windows created from mHealth data.")
        return None, None, None, None

    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(all_labels)
    X_windowed = np.array(all_windows, dtype=np.float32)

    X_windowed = np.transpose(X_windowed, (0, 2, 1))

    mhealth_activity_mapping = {
        1: 'Standing still', 2: 'Sitting and relaxing', 3: 'Lying down', 4: 'Walking',
        5: 'Climbing stairs', 6: 'Waist bends forward', 7: 'Frontal elevation of arms',
        8: 'Knees bending', 9: 'Cycling', 10: 'Jogging', 11: 'Running', 12: 'Jump front & back'
    }
    activity_labels = []
    class_names = list(label_encoder.classes_)

    for original_label in class_names:
        if original_label in mhealth_activity_mapping:
            activity_labels.append(mhealth_activity_mapping[original_label])
        else:
            activity_labels.append(f"Unknown_Activity_{original_label}")

    print("\n" + "=" * 50)
    print("Per-Subject Class Distribution (Window Count) - MHEALTH")
    print("=" * 50)

    sub_array = np.array(all_subjects, dtype=int)
    all_subjects_u = np.unique(sub_array)
    all_labels_idx = sorted(list(np.unique(y_encoded)))

    dist_data = []

    for sub_id in all_subjects_u:
        sub_mask = (sub_array == sub_id)
        sub_labels = y_encoded[sub_mask]
        class_counts = Counter(sub_labels)

        row = {'Subject': f"S{sub_id}"}
        total = 0
        for label_idx in all_labels_idx:
            count = class_counts.get(label_idx, 0)
            row[activity_labels[label_idx]] = count
            total += count
        row['Total'] = total
        dist_data.append(row)

    dist_df = pd.DataFrame(dist_data)
    dist_df = dist_df.set_index('Subject')

    col_order = [activity_labels[i] for i in all_labels_idx] + ['Total']
    dist_df = dist_df[col_order]

    display(dist_df)
    print("=" * 50 + "\n")

    return X_windowed, y_encoded, np.array(all_subjects, dtype=int), activity_labels

def load_full_uci_har_data(root, **kwargs):
    print(f"Loading UCI-HAR data from: {root}")
    pref = ["total_acc_x_","total_acc_y_","total_acc_z_",
            "body_acc_x_","body_acc_y_","body_acc_z_",
            "body_gyro_x_","body_gyro_y_","body_gyro_z_"]

    def load(s):
        print(f"  - Loading UCI-HAR '{s}' files...")

        inertial_path = os.path.join(root)
        main_path = os.path.join(root)

        if not os.path.isdir(inertial_path):
            raise FileNotFoundError(f"Inertial Signals path not found: {inertial_path}")
        if not os.path.isdir(main_path):
            raise FileNotFoundError(f"Main path not found: {main_path}")

        ch = [read_txt_matrix(os.path.join(inertial_path, f"{p}{s}.txt")) for p in pref]
        X = np.stack(ch, axis=1)

        y_path = os.path.join(main_path, f"y_{s}.txt")
        y = read_txt_matrix(y_path).astype(int).reshape(-1) - 1

        sub_path = os.path.join(main_path, f"subject_{s}.txt")
        sub = read_txt_matrix(sub_path).astype(int).reshape(-1)

        return X,y,sub

    Xtr,ytr,str = load("train")
    Xte,yte,ste = load("test")
    X = np.concatenate((Xtr,Xte))
    y = np.concatenate((ytr,yte))
    sub = np.concatenate((str,ste))

    X_flat = X.reshape(-1, X.shape[-1])
    scaler = StandardScaler()
    X_scaled_flat = scaler.fit_transform(X_flat.reshape(-1, 9))
    X = X_scaled_flat.reshape(X.shape)

    X = X.astype(np.float32)

    names = ['Walking','Walking Upstairs','Walking Downstairs','Sitting','Standing','Laying']

    print("\n" + "=" * 50)
    print("Per-Subject Class Distribution (Window Count) - UCI-HAR")
    print("=" * 50)

    all_subjects_u = np.unique(sub)
    all_labels_idx = sorted(list(np.unique(y)))

    dist_data = []

    for sub_id in all_subjects_u:
        sub_mask = (sub == sub_id)
        sub_labels = y[sub_mask]
        class_counts = Counter(sub_labels)

        row = {'Subject': f"S{sub_id}"}
        total = 0
        for label_idx in all_labels_idx:
            count = class_counts.get(label_idx, 0)
            row[names[label_idx]] = count
            total += count
        row['Total'] = total
        dist_data.append(row)

    dist_df = pd.DataFrame(dist_data)
    dist_df = dist_df.set_index('Subject')

    col_order = [names[i] for i in all_labels_idx] + ['Total']
    dist_df = dist_df[col_order]

    display(dist_df)
    print("=" * 50 + "\n")

    return X,y,sub,names

class CompactEncoder(nn.Module):
    def __init__(self, in_channels=9, embed_dim=128):
        super().__init__()
        self.stage1 = self._make_stage(in_channels, 32, stride=2)
        self.stage2 = self._make_stage(32, 64, stride=2)
        self.stage3 = self._make_stage(64, 128, stride=2)
        self.stage4 = self._make_stage(128, embed_dim, stride=2)
        self.pool = nn.AdaptiveAvgPool1d(1)

    def _make_stage(self, in_ch, out_ch, stride=1):
        return nn.Sequential(
            nn.Conv1d(in_ch, out_ch, kernel_size=5, stride=stride, padding=2),
            nn.BatchNorm1d(out_ch),
            nn.GELU(),
            nn.Conv1d(out_ch, out_ch, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(out_ch),
            nn.GELU(),
            nn.Conv1d(out_ch, out_ch, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm1d(out_ch),
            nn.GELU()
        )

    def forward(self, x):
        x = self.stage1(x)
        x = self.stage2(x)
        x = self.stage3(x)
        x = self.stage4(x)
        x = self.pool(x).flatten(1)
        return x

class AttentionWeightedProtoNet(nn.Module):
    def __init__(self, encoder, feature_dim, temperature=1.0):
        super().__init__()
        self.encoder = encoder
        self.temperature = temperature

        self.attention = nn.Sequential(
            nn.Linear(feature_dim, feature_dim // 2),
            nn.ReLU(),
            nn.Linear(feature_dim // 2, 1)
        )
        self.adapter = nn.Sequential(
            nn.Linear(feature_dim * 2, feature_dim),
            nn.ReLU(),
            nn.Linear(feature_dim, feature_dim)
        )

    def compute_prototypes(self, support_features, support_y, n_way):
        embed_dim = support_features.size(1)
        prototypes = []

        for c in range(n_way):
            mask = (support_y == c)
            if mask.sum() > 0:
                features_c = support_features[mask]

                attn_w = F.softmax(self.attention(features_c), dim=0)

                prototype_c = (attn_w * features_c).sum(0)

                prototypes.append(prototype_c)
            else:
                prototypes.append(torch.zeros(embed_dim, device=support_features.device))

        return torch.stack(prototypes)

    def forward(self, query_x, support_x, support_y, n_way):
        support_features = self.encoder(support_x)
        query_features = self.encoder(query_x)

        prototypes = self.compute_prototypes(support_features, support_y, n_way)

        n_query = query_features.size(0)

        query_features_expanded = query_features.unsqueeze(1).expand(-1, n_way, -1)
        proto_features_expanded = prototypes.unsqueeze(0).expand(n_query, -1, -1)

        cat_input = torch.cat([query_features_expanded, proto_features_expanded], dim=-1)
        delta = self.adapter(cat_input)
        adapted_prototypes = proto_features_expanded + delta

        dist = ((query_features_expanded - adapted_prototypes)**2).sum(dim=-1)

        logits = -dist / self.temperature

        return logits

class OriginalProtoNet(nn.Module):
    def __init__(self, encoder, feature_dim, temperature=1.0):
        super().__init__()
        self.encoder = encoder
        self.temperature = temperature

    def compute_prototypes(self, support_features, support_y, n_way):
        embed_dim = support_features.size(1)
        prototypes = []

        for c in range(n_way):
            mask = (support_y == c)
            if mask.sum() > 0:
                features_c = support_features[mask]
                prototype_c = features_c.mean(dim=0)
                prototypes.append(prototype_c)
            else:
                prototypes.append(torch.zeros(embed_dim, device=support_features.device))

        return torch.stack(prototypes)

    def forward(self, query_x, support_x, support_y, n_way):
        support_features = self.encoder(support_x)
        query_features = self.encoder(query_x)

        prototypes = self.compute_prototypes(support_features, support_y, n_way)

        n_query = query_features.size(0)

        query_features_expanded = query_features.unsqueeze(1).expand(-1, n_way, -1)
        proto_features_expanded = prototypes.unsqueeze(0).expand(n_query, -1, -1)

        dist = ((query_features_expanded - proto_features_expanded)**2).sum(dim=-1)

        logits = -dist / self.temperature

        return logits

class SupervisedBaseline(nn.Module):
    def __init__(self, in_channels=9, embed_dim=128, num_classes=6):
        super().__init__()
        self.encoder = CompactEncoder(in_channels=in_channels, embed_dim=embed_dim)
        self.classifier = nn.Linear(embed_dim, num_classes)

    def extract_features(self, x):
        return self.encoder(x)

    def forward(self, x):
        features = self.extract_features(x)
        return self.classifier(features)

def create_episodes(data_pool_list, n_way=6, k_shot=5, q_query=15, total_classes=6):

    data_pool = data_pool_list[0]
    data = data_pool['data']
    labels = data_pool['labels']
    unique_cls = np.unique(labels)

    for _ in range(50):
        if len(unique_cls) < n_way:
            continue

        chosen_classes = np.random.choice(unique_cls, n_way, replace=False)
        support_x, support_y, query_x, query_y = [], [], [], []

        valid_episode = True
        for new_c, original_c in enumerate(chosen_classes):
            idxs = np.where(labels == original_c)[0]
            if len(idxs) < k_shot + q_query:
                valid_episode = False
                break

            np.random.shuffle(idxs)
            s_idx = idxs[:k_shot]
            q_idx = idxs[k_shot:k_shot+q_query]

            support_x.extend([data[i] for i in s_idx])
            support_y.extend([new_c] * k_shot)
            query_x.extend([data[i] for i in q_idx])
            query_y.extend([new_c] * q_query)

        if valid_episode:
            if not support_x or not query_x:
                continue
            return (torch.stack(support_x), torch.tensor(support_y),
                    torch.stack(query_x), torch.tensor(query_y))

    return (torch.empty(0), torch.empty(0), torch.empty(0), torch.empty(0))

def meta_train(model, optimizer, train_users, val_users, n_way_range, k_shot=5, q_query=15,
               episodes=5000, log_every=100, val_every=200, total_classes=6):
    model.train()
    train_losses, train_accs, val_accs = [], [], []

    augmentation_funcs = [
        aug_jitter_scale,
        aug_sensor_failure,
        aug_missing_segment,
        aug_outlier_spike
    ]

    aug_plan = []

    n_clean = int(episodes * 0.5)
    n_aug_per_type = int(episodes * 0.5) // len(augmentation_funcs)

    aug_plan.extend([None] * n_clean)
    for func in augmentation_funcs:
        aug_plan.extend([func] * n_aug_per_type)

    remainder = episodes - len(aug_plan)
    aug_plan.extend([None] * remainder)

    random.shuffle(aug_plan)

    print(f"Training Strategy: Fully shuffled list of {len(aug_plan)} episodes.")
    print(f"(Composition: {n_clean + remainder} Clean, 4x{n_aug_per_type} Augmentations)")

    for ep in range(episodes):
        n_way = random.randint(n_way_range[0], n_way_range[1])

        support_x, support_y, query_x, query_y = create_episodes(train_users, n_way, k_shot, q_query, total_classes=total_classes)
        if support_x.size(0) == 0:
            continue

        if ep < len(aug_plan):
            aug_func = aug_plan[ep]
        else:
            aug_func = None

        if aug_func is None:
            support_x_aug = support_x
            query_x_aug = query_x
        else:
            support_x_aug = aug_func(support_x)
            query_x_aug = aug_func(query_x)

        support_x_aug, support_y = support_x_aug.cuda(), support_y.cuda()
        query_x_aug, query_y = query_x_aug.cuda(), query_y.cuda()

        logits = model(query_x_aug, support_x_aug, support_y, n_way)
        loss = F.cross_entropy(logits, query_y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        acc = (logits.argmax(1) == query_y).float().mean().item()
        train_losses.append(loss.item())
        train_accs.append(acc)

        if (ep + 1) % log_every == 0:
            print(f"Episode {ep+1:5d} | Loss: {loss.item():.4f} | Train Acc: {acc:.4f}", end="")
            if (ep + 1) % val_every == 0:
                val_acc = meta_validate(model, val_users, k_shot, n_way_range[1], trials=200, total_classes=total_classes)
                val_accs.append(val_acc)
                print(f" | Val Acc: {val_acc:.4f}")
            else:
                print()

    val_episodes = list(range(val_every - 1, episodes, val_every))
    if not val_episodes:
        val_episodes = [episodes - 1] if episodes > 0 else [0]
    if not val_accs:
        val_accs = [0.0]

    val_accs_interpolated = np.interp(range(episodes), val_episodes, val_accs)

    return train_losses, train_accs, val_accs_interpolated

def meta_validate(model, val_users, k_shot=5, n_way=6, trials=20, total_classes=6):
    model.eval()
    correct, total = 0, 0
    for _ in range(trials):
        support_x, support_y, query_x, query_y = create_episodes(val_users, n_way, k_shot, 15, total_classes=total_classes)
        if support_x.size(0) == 0:
            continue
        support_x, support_y = support_x.cuda(), support_y.cuda()
        query_x, query_y = query_x.cuda(), query_y.cuda()
        with torch.no_grad():
            logits = model(query_x, support_x, support_y, n_way)
            preds = logits.argmax(1)
            correct += (preds == query_y).sum().item()
            total += len(query_y)
    model.train()
    return correct / total if total > 0 else 0.0

def meta_test(model, test_users, k_shot=5, n_way=6, q_query=15, trials=100, total_classes=6):
    model.eval()
    all_preds, all_trues = [], []

    for _ in range(trials):
        support_x, support_y, query_x, query_y = create_episodes(test_users, n_way, k_shot, q_query, total_classes=total_classes)
        if support_x.size(0) == 0:
            continue
        support_x, support_y = support_x.cuda(), support_y.cuda()
        query_x, query_y = query_x.cuda(), query_y.cuda()

        with torch.no_grad():
            logits = model(query_x, support_x, support_y, n_way)
            preds = logits.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_trues.extend(query_y.cpu().numpy())

    if not all_trues:
        return 0, 0, 0, 0, np.zeros((n_way, n_way))

    acc = accuracy_score(all_trues, all_preds)
    f1 = f1_score(all_trues, all_preds, average='macro', zero_division=0)
    prec = precision_score(all_trues, all_preds, average='macro', zero_division=0)
    rec = recall_score(all_trues, all_preds, average='macro', zero_division=0)
    cm = confusion_matrix(all_trues, all_preds, labels=range(n_way))

    return acc, f1, prec, rec, cm

def meta_test_noisy(model, test_users, k_shot=5, n_way=6, q_query=15, trials=100, noise_type='jitter', noise_level=0.1, total_classes=6):
    model.eval()
    all_preds, all_trues = [], []

    for _ in range(trials):
        support_x, support_y, query_x, query_y = create_episodes(test_users, n_way, k_shot, q_query, total_classes=total_classes)
        if support_x.size(0) == 0:
            continue
        support_x, support_y = support_x.cuda(), support_y.cuda()
        query_x, query_y = query_x.cuda(), query_y.cuda()

        query_x_noisy = apply_test_noise(query_x, noise_type, noise_level)

        with torch.no_grad():
            logits = model(query_x_noisy, support_x, support_y, n_way)
            preds = logits.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_trues.extend(query_y.cpu().numpy())

    if not all_trues:
        return 0, 0, 0, 0, np.zeros((n_way, n_way))

    acc = accuracy_score(all_trues, all_preds)
    f1 = f1_score(all_trues, all_preds, average='macro', zero_division=0)
    prec = precision_score(all_trues, all_preds, average='macro', zero_division=0)
    rec = recall_score(all_trues, all_preds, average='macro', zero_division=0)
    cm = confusion_matrix(all_trues, all_preds, labels=range(n_way))

    return acc, f1, prec, rec, cm

def train_supervised(model, train_loader, optimizer, criterion, epochs, device):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        total_correct = 0
        total_samples = 0
        for x_batch, y_batch in train_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)

            optimizer.zero_grad()
            logits = model(x_batch)
            loss = criterion(logits, y_batch)
            loss.backward()
            optimizer.step()

            total_loss += loss.item() * x_batch.size(0)
            total_correct += (logits.argmax(1) == y_batch).sum().item()
            total_samples += x_batch.size(0)

        if total_samples == 0:
            avg_loss, avg_acc = 0, 0
        else:
            avg_loss = total_loss / total_samples
            avg_acc = total_correct / total_samples

        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1:3d}/{epochs} | Train Loss: {avg_loss:.4f} | Train Acc: {avg_acc:.4f}")

def test_supervised(model, test_loader, device, num_classes=6):
    model.eval()
    all_preds, all_trues = [], []
    with torch.no_grad():
        for x_batch, y_batch in test_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            logits = model(x_batch)
            preds = logits.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_trues.extend(y_batch.cpu().numpy())

    if not all_trues:
        return 0, 0, 0, 0, np.zeros((num_classes, num_classes))

    acc = accuracy_score(all_trues, all_preds)
    f1 = f1_score(all_trues, all_preds, average='macro', zero_division=0)
    prec = precision_score(all_trues, all_preds, average='macro', zero_division=0)
    rec = recall_score(all_trues, all_preds, average='macro', zero_division=0)
    cm = confusion_matrix(all_trues, all_preds, labels=range(num_classes))
    return acc, f1, prec, rec, cm

def test_supervised_noisy(model, test_loader, device, noise_type, noise_level, num_classes=6):
    model.eval()
    all_preds, all_trues = [], []
    with torch.no_grad():
        for x_batch, y_batch in test_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            x_batch_noisy = apply_test_noise(x_batch, noise_type, noise_level)
            logits = model(x_batch_noisy)
            preds = logits.argmax(1).cpu().numpy()
            all_preds.extend(preds)
            all_trues.extend(y_batch.cpu().numpy())

    if not all_trues:
        return 0, 0, 0, 0, np.zeros((num_classes, num_classes))

    acc = accuracy_score(all_trues, all_preds)
    f1 = f1_score(all_trues, all_preds, average='macro', zero_division=0)
    prec = precision_score(all_trues, all_preds, average='macro', zero_division=0)
    rec = recall_score(all_trues, all_preds, average='macro', zero_division=0)
    cm = confusion_matrix(all_trues, all_preds, labels=range(num_classes))
    return acc, f1, prec, rec, cm

def prepare_users(X, y, sub, subs, mean, std):
    users = []
    for uid in subs:
        mask = (sub == uid)
        user_data = X[mask]
        user_labels = y[mask]
        if len(user_data) == 0:
            continue
        user_dataset = HARDataset(user_data, user_labels, norm=False)
        user_dataset.set_stats(mean, std)

        if len(user_dataset) > 0:
            users.append({
                'data': torch.stack([user_dataset[i][0] for i in range(len(user_dataset))]),
                'labels': user_labels,
                'id': uid
            })
    return users

if __name__ == '__main__':
    set_seed(42)
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"Device: {device}\n")

    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', 1000)

    UCI_ROOT = "/content/drive/MyDrive/HAR_Dataset/UCI"
    MHEALTH_ROOT = "/content/drive/MyDrive/HAR_Dataset/MHEALTH"
    MOBIACT_ROOT = "/content/drive/MyDrive/HAR_Dataset/MOBIACT"
    OPPORTUNITY_ROOT = "/content/drive/MyDrive/HAR_Dataset/OPPORTUNITY"


    dataset_definitions = [
        {
            'name': 'MobiAct',
            'loader': load_mobiact_data,
            'path': MOBIACT_ROOT,
            'params': {'w_s': 100, 'stride': 50}
        },
        {
            'name': 'OPPORTUNITY',
            'loader': load_opportunity_local_data,
            'path': OPPORTUNITY_ROOT,
            'params': {'window_size': 30, 'step': 15}
        },
        {
            'name': 'UCI-HAR',
            'loader': load_full_uci_har_data,
            'path': UCI_ROOT,
            'params': {}
        },
        {
            'name': 'MHEALTH',
            'loader': load_mhealth_data,
            'path': MHEALTH_ROOT,
            'params': {}
        },
    ]

    noise_conditions_to_test = [
        ('jitter', 0.1), ('jitter', 0.2), ('jitter', 0.3),
        ('time_shift', 0.1), ('time_shift', 0.2), ('time_shift', 0.3),
        ('sensor_failure', 0.1), ('sensor_failure', 0.2), ('sensor_failure', 0.3),
        ('outlier_spike', 0.1), ('outlier_spike', 0.2), ('outlier_spike', 0.3),
        ('missing_segment', 0.1), ('missing_segment', 0.2), ('missing_segment', 0.3)
    ]
    train_k_shots_list = [1, 3, 5, 10]
    meta_models_to_test = [
        ('OriginalProtoNet', OriginalProtoNet),
        ('AttentionWeightedProtoNet', AttentionWeightedProtoNet)
    ]

    EPISODES = 10000
    N_WAY_TRAIN_MIN = 3
    Q_QUERY_TRAIN = 15

    print("\n" + "#" * 90)
    print("##### STAGE 1: PRE-LOADING AND PRE-PROCESSING ALL DATASETS #####")
    print("#" * 90 + "\n")

    loaded_data_cache = {}

    for dataset_info in dataset_definitions:
        dataset_name = dataset_info['name']
        loader_func = dataset_info['loader']
        dataset_path = dataset_info['path']
        loader_params = dataset_info['params']

        print("\n" + "=" * 80)
        print(f"LOADING {dataset_name} DATASET from {dataset_path}")
        print("=" * 80)

        X_raw, y_raw, sub_raw, labels = loader_func(dataset_path, **loader_params)

        if X_raw is None:
            print(f"Failed to load {dataset_name} data. Skipping...")
            continue

        print(f"Data shape (N, C, L): {X_raw.shape}")
        print(f"Labels shape: {y_raw.shape}")
        print(f"Subjects shape: {sub_raw.shape}")
        print(f"Total classes: {len(labels)}")
        print(f"Class names: {labels}")

        loaded_data_cache[dataset_name] = {
            'X_raw': X_raw,
            'y_raw': y_raw,
            'sub_raw': sub_raw,
            'labels': labels
        }
        print(f"--- Finished loading and caching {dataset_name} ---")

    print("\n" + "#" * 90)
    print("##### STAGE 2: RUNNING EXPERIMENTS ON LOADED DATA #####")
    print("#" * 90 + "\n")

    for dataset_name, data_cache in loaded_data_cache.items():

        print("\n" + "#" * 90)
        print(f"##### STARTING EXPERIMENT FOR DATASET: {dataset_name} #####")
        print("#" * 90 + "\n")

        X_raw = data_cache['X_raw']
        y_raw = data_cache['y_raw']
        sub_raw = data_cache['sub_raw']
        labels = data_cache['labels']

        unique_subs = np.unique(sub_raw)
        print(f"Total subjects: {len(unique_subs)} (IDs: {unique_subs})")
        print(f"Using CLASS-STRATIFIED split (Train/Val/Test = 70/10/20)")

        INPUT_CHANNELS = X_raw.shape[1]
        INPUT_LENGTH = X_raw.shape[2]
        TOTAL_CLASSES = len(labels)
        N_WAY_TRAIN_MAX = TOTAL_CLASSES

        print(f"Input Channels: {INPUT_CHANNELS}, Input Length: {INPUT_LENGTH}, Total Classes: {TOTAL_CLASSES}")

        current_dataset_results = []
        current_dataset_stats = []

        print("\n" + "*" * 80)
        print(f"STARTING SINGLE RUN (Dataset: {dataset_name})")
        print("Splitting data into stratified Train(70%)/Val(10%)/Test(20%)...")
        print("*" * 80)

        X_train_val, X_test, y_train_val, y_test = train_test_split(
            X_raw, y_raw,
            test_size=0.2,
            random_state=42,
            stratify=y_raw
        )

        val_split_ratio = 0.1 / 0.8

        X_train, X_val, y_train, y_val = train_test_split(
            X_train_val, y_train_val,
            test_size=val_split_ratio,
            random_state=42,
            stratify=y_train_val
        )

        if len(X_train) == 0 or len(X_val) == 0 or len(X_test) == 0:
            print(f"Dataset {dataset_name} has empty data splits after stratification. Skipping.")
            continue

        print(f"Train samples: {len(X_train)}")
        print(f"Val samples:   {len(X_val)}")
        print(f"Test samples:  {len(X_test)}")

        mean, std = compute_mean_std(X_train)

        train_dataset_sup = HARDataset(X_train, y_train, norm=False)
        train_dataset_sup.set_stats(mean, std)
        train_loader_sup = DataLoader(train_dataset_sup, batch_size=512, shuffle=True)

        test_dataset_sup = HARDataset(X_test, y_test, norm=False)
        test_dataset_sup.set_stats(mean, std)
        test_loader_sup = DataLoader(test_dataset_sup, batch_size=512, shuffle=False)

        X_train_norm = (X_train - mean) / std
        X_val_norm = (X_val - mean) / std
        X_test_norm = (X_test - mean) / std

        train_users = [{'data': torch.from_numpy(X_train_norm.astype(np.float32)), 'labels': y_train, 'id': 'train_pool'}]
        val_users = [{'data': torch.from_numpy(X_val_norm.astype(np.float32)), 'labels': y_val, 'id': 'val_pool'}]
        test_users = [{'data': torch.from_numpy(X_test_norm.astype(np.float32)), 'labels': y_test, 'id': 'test_pool'}]

        if not train_users or not val_users or not test_users:
            print(f"Dataset {dataset_name} resulted in empty user lists after processing. Skipping.")
            continue


        print("\n" + "=" * 80)
        print(f"PART 1: SUPERVISED BASELINE")
        print("=" * 80)

        supervised_model = SupervisedBaseline(
            in_channels=INPUT_CHANNELS,
            embed_dim=128,
            num_classes=TOTAL_CLASSES
        ).to(device)

        optimizer_sup = torch.optim.Adam(supervised_model.parameters(), lr=0.001)
        criterion_sup = nn.CrossEntropyLoss()

        print("--- Training Supervised Baseline ---")
        train_supervised(supervised_model, train_loader_sup, optimizer_sup, criterion_sup, epochs=50, device=device)

        print("\n--- Calculating Supervised Model Stats ---")
        sup_params, sup_flops = compute_supervised_stats(supervised_model, input_shape=(INPUT_CHANNELS, INPUT_LENGTH))
        sup_inf_time = measure_supervised_inference_time(supervised_model, input_size=(1, INPUT_CHANNELS, INPUT_LENGTH), device=device)
        print(f"Supervised Stats | Params(M): {sup_params:.2f}, FLOPs(M): {sup_flops:.2f}, Inf. Time(ms): {sup_inf_time:.2f}")

        current_dataset_stats.append({
            'Model': 'Supervised',
            'K-Shot': 'N/A (Full)',
            'Param (M)': sup_params,
            'FLOPs (M)': sup_flops,
            'Inf. Time (ms)': sup_inf_time
        })

        print("\n--- Testing Supervised Baseline (Clean) ---")
        acc_s, f1_s, prec_s, rec_s, _ = test_supervised(supervised_model, test_loader_sup, device, num_classes=TOTAL_CLASSES)
        print(f"Supervised Clean | Acc: {acc_s:.4f}, F1: {f1_s:.4f}")
        current_dataset_results.append({
            'Model': 'Supervised',
            'K-Shot': 'N/A (Full)',
            'Noise': 'Clean',
            'Accuracy': acc_s,
            'F1-Score': f1_s,
            'Precision': prec_s,
            'Recall': rec_s
        })

        print("\n--- Testing Supervised Baseline (Robustness) ---")
        for n_type, n_level in noise_conditions_to_test:
            acc_n, f1_n, prec_n, rec_n, _ = test_supervised_noisy(
                supervised_model, test_loader_sup, device,
                noise_type=n_type, noise_level=n_level, num_classes=TOTAL_CLASSES
            )
            print(f"Supervised Noisy | Type: {n_type:16s} | Lvl: {n_level} | Acc: {acc_n:.4f}, F1: {f1_n:.4f}")
            current_dataset_results.append({
                'Model': 'Supervised',
                'K-Shot': 'N/A (Full)',
                'Noise': f"{n_type}_{n_level}",
                'Accuracy': acc_n,
                'F1-Score': f1_n,
                'Precision': prec_n,
                'Recall': rec_n
            })

        del supervised_model, optimizer_sup, criterion_sup
        torch.cuda.empty_cache()

        print("\n" + "=" * 80)
        print(f"PART 2: META-LEARNING (Per K-Shot Training)")
        print("=" * 80)

        for model_name, model_class in meta_models_to_test:

            print("\n" + "=" * 80)
            print(f"===== STARTING META-TRAINING FOR: {model_name} =====")
            print("=" * 80)

            for k_train in train_k_shots_list:
                print("\n" + "=" * 70)
                print(f"STARTING {model_name} FOR K-SHOT = {k_train}")
                print("=" * 70)

                K_SHOT_TRAIN = k_train

                encoder = CompactEncoder(in_channels=INPUT_CHANNELS, embed_dim=128).to(device)
                meta_model = model_class(encoder, feature_dim=128, temperature=1.0).to(device)
                optimizer_meta = torch.optim.Adam(meta_model.parameters(), lr=0.001)

                print(f"--- Calculating Meta Model Stats (Model={model_name}, K={k_train}) ---")
                meta_enc_params, meta_enc_flops = compute_encoder_stats(encoder, input_shape=(1, INPUT_CHANNELS, INPUT_LENGTH))
                meta_e2e_inf_time = measure_meta_inference_time(meta_model, n_way=TOTAL_CLASSES, sz=(1, INPUT_CHANNELS, INPUT_LENGTH), dev=device)
                print(f"Stats-{model_name}-{k_train} | Encoder Params(M): {meta_enc_params:.2f}, Encoder FLOPs(M): {meta_enc_flops:.2f}, E2E Inf. Time(ms): {meta_e2e_inf_time:.2f}")

                current_dataset_stats.append({
                    'Model': model_name,
                    'K-Shot': k_train,
                    'Param (M)': meta_enc_params,
                    'FLOPs (M)': meta_enc_flops,
                    'Inf. Time (ms)': meta_e2e_inf_time
                })

                print(f"--- Meta-Training (Model={model_name}, K={K_SHOT_TRAIN}, {EPISODES} Episodes) ---")
                train_losses, train_accs, val_accs = meta_train(
                    meta_model, optimizer_meta, train_users, val_users,
                    n_way_range=(N_WAY_TRAIN_MIN, N_WAY_TRAIN_MAX),
                    k_shot=K_SHOT_TRAIN,
                    q_query=Q_QUERY_TRAIN,
                    episodes=EPISODES,
                    log_every=500,
                    val_every=500,
                    total_classes=TOTAL_CLASSES
                )

                print(f"\n--- Testing Meta-Learning Model (Model={model_name}, K={k_train}) ---")

                print(f"Test-{model_name}-{k_train}-Shot (Clean) | Testing...")
                acc_m, f1_m, prec_m, rec_m, _ = meta_test(
                    meta_model, test_users, k_shot=k_train, n_way=TOTAL_CLASSES,
                    q_query=15, trials=250, total_classes=TOTAL_CLASSES
                )
                print(f"Test-{model_name}-{k_train}-Shot (Clean) | Acc: {acc_m:.4f}, F1: {f1_m:.4f}")
                current_dataset_results.append({
                    'Model': model_name,
                    'K-Shot': k_train,
                    'Noise': 'Clean',
                    'Accuracy': acc_m,
                    'F1-Score': f1_m,
                    'Precision': prec_m,
                    'Recall': rec_m
                })

                for n_type, n_level in noise_conditions_to_test:
                    acc_n, f1_n, prec_n, rec_n, _ = meta_test_noisy(
                        meta_model, test_users, k_shot=k_train, n_way=TOTAL_CLASSES, q_query=15,
                        trials=400, noise_type=n_type, noise_level=n_level,
                        total_classes=TOTAL_CLASSES
                    )
                    print(f"Test-{model_name}-{k_train}-Shot Noisy | Type: {n_type:16s} | Lvl: {n_level} | Acc: {acc_n:.4f}, F1: {f1_n:.4f}")
                    current_dataset_results.append({
                        'Model': model_name,
                        'K-Shot': k_train,
                        'Noise': f"{n_type}_{n_level}",
                        'Accuracy': acc_n,
                        'F1-Score': f1_n,
                        'Precision': prec_n,
                        'Recall': rec_n
                    })

                del encoder, meta_model, optimizer_meta
                torch.cuda.empty_cache()

        if not current_dataset_results:
            print(f"No results generated for dataset {dataset_name}. Skipping aggregation.")
            continue

        print("\n" + "=" * 80)
        print(f"===== FINAL AGGREGATED RESULTS FOR: {dataset_name} =====")
        print(f"(Based on a single class-stratified split)")
        print("=" * 80)

        final_df = pd.DataFrame(current_dataset_results)

        metric_cols = ['Accuracy', 'F1-Score', 'Precision', 'Recall']
        display_df = final_df.copy()

        for col in metric_cols:
            display_df[col] = display_df[col].apply(lambda x: f"{x:.4f}")

        display(display_df)

        print("\n" + "=" * 80)
        print(f"===== PIVOT TABLE: ACCURACY ({dataset_name}) =====")
        print("=" * 80)

        pivot_acc_str = display_df.pivot_table(
            index='Noise',
            columns=['Model', 'K-Shot'],
            values='Accuracy',
            aggfunc='first'
        )

        cols_order = [
            ('Supervised', 'N/A (Full)'),
            ('OriginalProtoNet', 1),
            ('OriginalProtoNet', 3),
            ('OriginalProtoNet', 5),
            ('AttentionWeightedProtoNet', 1),
            ('AttentionWeightedProtoNet', 3),
            ('AttentionWeightedProtoNet', 5)
        ]

        cols_present = [c for c in cols_order if c in pivot_acc_str.columns]
        pivot_acc_str = pivot_acc_str[cols_present]
        display(pivot_acc_str)


        print("\n" + "=" * 80)
        print(f"===== PIVOT TABLE: F1-SCORE ({dataset_name}) =====")
        print("=" * 80)

        pivot_f1_str = display_df.pivot_table(
            index='Noise',
            columns=['Model', 'K-Shot'],
            values='F1-Score',
            aggfunc='first'
        )

        cols_present_f1 = [c for c in cols_order if c in pivot_f1_str.columns]
        pivot_f1_str = pivot_f1_str[cols_present_f1]
        display(pivot_f1_str)

        print("\n" + "=" * 80)
        print(f"===== MODEL STATISTICS ({dataset_name}) =====")
        print("=" * 80)

        stats_df = pd.DataFrame(current_dataset_stats)
        stats_cols = ['Param (M)', 'FLOPs (M)', 'Inf. Time (ms)']

        display_stats_df = stats_df.copy()
        for col in stats_cols:
             display_stats_df[col] = display_stats_df[col].apply(lambda x: f"{x:.2f}")

        display(display_stats_df.set_index(['Model', 'K-Shot']))


        print("\n" + "=" * 80)
        print(f"===== K-SHOT PERFORMANCE COMPARISON (F1-SCORE) ({dataset_name}) =====")
        print("=" * 80)

        df_clean = final_df[final_df['Noise'] == 'Clean'].copy()

        f1_supervised = 0.0
        sup_data = df_clean[df_clean['Model'] == 'Supervised']
        if not sup_data.empty:
            f1_supervised = sup_data['F1-Score'].values[0]

        k_shots_numeric = sorted(df_clean[df_clean['Model'] != 'Supervised']['K-Shot'].unique())
        if not k_shots_numeric:
            k_shots_numeric = [1, 3, 5]

        k_shots_present = [k for k in train_k_shots_list if k in k_shots_numeric]

        df_sup_line = pd.DataFrame({
            'Model': 'Supervised',
            'K-Shot': k_shots_present,
            'F1-Score': f1_supervised
        })

        df_meta = df_clean[df_clean['Model'] != 'Supervised']
        plot_df = pd.concat([df_meta, df_sup_line])

        plt.figure(figsize=(7, 4.5))

        models_to_plot = plot_df['Model'].unique()

        for model in models_to_plot:
            model_df = plot_df[plot_df['Model'] == model]
            plt.plot(
                model_df['K-Shot'],
                model_df['F1-Score'],
                label=model,
                marker='o',
                markersize=8,
                linewidth=2
            )

        plt.title(f'{dataset_name}: Model F1-Score vs. K-Shot (Clean Data)', fontsize=14)
        plt.xlabel('K-Shot', fontsize=12)
        plt.ylabel('F1-Score', fontsize=12)
        plt.xticks(k_shots_present)
        plt.grid(True, linestyle='--', alpha=0.4)
        plt.legend(title='Model')
        plt.tight_layout()
        plt.show()

    print("\n" + "#" * 90)
    print("##### ALL DATASET EXPERIMENTS COMPLETE #####")
    print("#" * 90)
```
