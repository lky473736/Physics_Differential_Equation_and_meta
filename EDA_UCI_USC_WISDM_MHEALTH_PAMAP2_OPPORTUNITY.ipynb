```python
import os
import glob
import random
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import scipy.io
from collections import Counter
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
from IPython.display import display, clear_output

def set_seed(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (15, 7)
plt.rcParams['figure.autolayout'] = True

BASE_PATH = "/content/drive/MyDrive/HAR_Dataset"
UCI_PATH = os.path.join(BASE_PATH, "UCI")
USC_PATH = os.path.join(BASE_PATH, "USC")
WISDM_PATH = os.path.join(BASE_PATH, "WISDM")
MHEALTH_PATH = os.path.join(BASE_PATH, "MHEALTH")
PAMAP2_PATH = os.path.join(BASE_PATH, "PAMAP2")
OPPORTUNITY_PATH = os.path.join(BASE_PATH, "OPPORTUNITY")

MHEALTH_COLS = [
    'acc_chest_x', 'acc_chest_y', 'acc_chest_z',
    'ecg_1', 'ecg_2',
    'acc_left_ankle_x', 'acc_left_ankle_y', 'acc_left_ankle_z',
    'gyro_left_ankle_x', 'gyro_left_ankle_y', 'gyro_left_ankle_z',
    'mag_left_ankle_x', 'mag_left_ankle_y', 'mag_left_ankle_z',
    'acc_right_lower_arm_x', 'acc_right_lower_arm_y', 'acc_right_lower_arm_z',
    'gyro_right_lower_arm_x', 'gyro_right_lower_arm_y', 'gyro_right_lower_arm_z',
    'mag_right_lower_arm_x', 'mag_right_lower_arm_y', 'mag_right_lower_arm_z',
    'label'
]

MHEALTH_ACTIVITY_MAPPING = {
    # 0: 'Null',
    1: 'Standing still',
    2: 'Sitting and relaxing',
    3: 'Lying down',
    4: 'Walking',
    5: 'Climbing stairs',
    6: 'Waist bends forward',
    7: 'Frontal elevation of arms',
    8: 'Knees bending (crouching)',
    9: 'Cycling',
    10: 'Jogging',
    11: 'Running',
    12: 'Jump front & back'
}

PAMAP2_COLS = [
    'timestamp', 'label', 'heart_rate', 'hand_temp',
    'hand_acc_16g_x', 'hand_acc_16g_y', 'hand_acc_16g_z',
    'hand_acc_6g_x', 'hand_acc_6g_y', 'hand_acc_6g_z',
    'hand_gyro_x', 'hand_gyro_y', 'hand_gyro_z',
    'hand_mag_x', 'hand_mag_y', 'hand_mag_z',
    'hand_orient_1', 'hand_orient_2', 'hand_orient_3', 'hand_orient_4',
    'chest_temp',
    'chest_acc_16g_x', 'chest_acc_16g_y', 'chest_acc_16g_z',
    'chest_acc_6g_x', 'chest_acc_6g_y', 'chest_acc_6g_z',
    'chest_gyro_x', 'chest_gyro_y', 'chest_gyro_z',
    'chest_mag_x', 'chest_mag_y', 'chest_mag_z',
    'chest_orient_1', 'chest_orient_2', 'chest_orient_3', 'chest_orient_4',
    'ankle_temp',
    'ankle_acc_16g_x', 'ankle_acc_16g_y', 'ankle_acc_16g_z',
    'ankle_acc_6g_x', 'ankle_acc_6g_y', 'ankle_acc_6g_z',
    'ankle_gyro_x', 'ankle_gyro_y', 'ankle_gyro_z',
    'ankle_mag_x', 'ankle_mag_y', 'ankle_mag_z',
    'ankle_orient_1', 'ankle_orient_2', 'ankle_orient_3', 'ankle_orient_4'
]

PAMAP2_FEATURE_COLS = [
    'hand_temp',
    'hand_acc_16g_x', 'hand_acc_16g_y', 'hand_acc_16g_z',
    'hand_acc_6g_x', 'hand_acc_6g_y', 'hand_acc_6g_z',
    'hand_gyro_x', 'hand_gyro_y', 'hand_gyro_z',
    'hand_mag_x', 'hand_mag_y', 'hand_mag_z',
    'hand_orient_1', 'hand_orient_2', 'hand_orient_3', 'hand_orient_4',
    'chest_temp',
    'chest_acc_16g_x', 'chest_acc_16g_y', 'chest_acc_16g_z',
    'chest_acc_6g_x', 'chest_acc_6g_y', 'chest_acc_6g_z',
    'chest_gyro_x', 'chest_gyro_y', 'chest_gyro_z',
    'chest_mag_x', 'chest_mag_y', 'chest_mag_z',
    'chest_orient_1', 'chest_orient_2', 'chest_orient_3', 'chest_orient_4',
    'ankle_temp',
    'ankle_acc_16g_x', 'ankle_acc_16g_y', 'ankle_acc_16g_z',
    'ankle_acc_6g_x', 'ankle_acc_6g_y', 'ankle_acc_6g_z',
    'ankle_gyro_x', 'ankle_gyro_y', 'ankle_gyro_z',
    'ankle_mag_x', 'ankle_mag_y', 'ankle_mag_z',
    'ankle_orient_1', 'ankle_orient_2', 'ankle_orient_3', 'ankle_orient_4'
]

PAMAP2_ACTIVITY_MAPPING = {
    # 0: 'Other',
    1: 'Lying',
    2: 'Sitting',
    3: 'Standing',
    4: 'Walking',
    5: 'Running',
    6: 'Cycling',
    7: 'Nordic walking',
    9: 'Watching TV',
    10: 'Computer work',
    11: 'Car driving',
    12: 'Ascending stairs',
    13: 'Descending stairs',
    16: 'Vacuum cleaning',
    17: 'Ironing',
    18: 'Folding laundry',
    19: 'House cleaning',
    20: 'Playing soccer',
    24: 'Rope jumping'
}

OPPORTUNITY_FEATURE_COLS = [
    'RKN^_acc_x', 'RKN^_acc_y', 'RKN^_acc_z',
    'RKN^_gyro_x', 'RKN^_gyro_y', 'RKN^_gyro_z',
    'HIP_acc_x', 'HIP_acc_y', 'HIP_acc_z',
    'HIP_gyro_x', 'HIP_gyro_y', 'HIP_gyro_z',
    'LUA_acc_x', 'LUA_acc_y', 'LUA_acc_z',
    'LUA_gyro_x', 'LUA_gyro_y', 'LUA_gyro_z',
    'RUA_acc_x', 'RUA_acc_y', 'RUA_acc_z',
    'RUA_gyro_x', 'RUA_gyro_y', 'RUA_gyro_z',
    'LH_acc_x', 'LH_acc_y', 'LH_acc_z',
    'LH_gyro_x', 'LH_gyro_y', 'LH_gyro_z'
]

OPPORTUNITY_COL_IDX = (
    [38, 39, 40, 41, 42, 43] +
    [51, 52, 53, 54, 55, 56] +
    [64, 65, 66, 67, 68, 69] +
    [77, 78, 79, 80, 81, 82] +
    [90, 91, 92, 93, 94, 95]
)
OPPORTUNITY_LABEL_COL = 249

OPPORTUNITY_ACTIVITY_MAPPING = {
    # 0: 'Null',
    406516: 'Open Door 1', 406517: 'Open Door 2',
    404516: 'Close Door 1', 404517: 'Close Door 2',
    406520: 'Open Fridge', 404520: 'Close Fridge',
    406505: 'Open Dishwasher', 404505: 'Close Dishwasher',
    406519: 'Open Drawer 1', 404519: 'Close Drawer 1',
    406511: 'Open Drawer 2', 404511: 'Close Drawer 2',
    406508: 'Open Drawer 3', 404508: 'Close Drawer 3',
    408512: 'Clean Table', 407521: 'Drink from Cup',
    405506: 'Toggle Switch'
}

UCI_ACTIVITY_MAPPING = {
    0: 'Walking',
    1: 'Walking Upstairs',
    2: 'Walking Downstairs',
    3: 'Sitting',
    4: 'Standing',
    5: 'Laying'
}

USC_ACTIVITY_MAPPING = {
    1: "Walking Forward", 2: "Walking Left", 3: "Walking Right",
    4: "Walking Upstairs", 5: "Walking Downstairs", 6: "Running Forward",
    7: "Jumping Up", 8: "Sitting", 9: "Standing",
    10: "Sleeping", 11: "Elevator Up", 12: "Elevator Down"
}

WISDM_ACTIVITY_MAPPING = {
    'Walking': 'Walking', 'Jogging': 'Jogging', 'Upstairs': 'Upstairs',
    'Downstairs': 'Downstairs', 'Sitting': 'Sitting', 'Standing': 'Standing'
}

def print_break(name):
    print("\n" + "="*80)
    print(f" {name} ")
    print("="*80 + "\n")

def print_file_info(dataset_dir, name):
    print_break(f"{name} - File Structure")
    file_count = 0
    total_size = 0
    for root, dirs, files in os.walk(dataset_dir):
        for f in files:
            if file_count < 20:
                print(f"  - {os.path.join(root, f).replace(BASE_PATH, '...')}")
            file_path = os.path.join(root, f)
            total_size += os.path.getsize(file_path)
            file_count += 1
    if file_count > 20:
        print(f"  ... and {file_count - 20} more files.")
    print(f"\nTotal Files: {file_count}")
    print(f"Total Size: {total_size / (1024*1024):.2f} MB")

def plot_missing_values(df, name):
    print_break(f"{name} - Missing Values Analysis")
    missing_counts = df.isnull().sum()
    missing_counts = missing_counts[missing_counts > 0]
    if missing_counts.empty:
        print("No missing values found. ")
        return

    missing_percent = (missing_counts / len(df)) * 100
    missing_df = pd.DataFrame({'count': missing_counts, 'percent': missing_percent})
    print(missing_df.sort_values(by='percent', ascending=False))

    plt.figure(figsize=(15, 7))
    missing_percent.plot(kind='bar')
    plt.title(f'Percentage of Missing Values by Feature - {name}')
    plt.ylabel('Percentage (%)')
    plt.xlabel('Features')
    plt.show()

def plot_class_distribution(df, label_col, name, activity_map=None):
    print_break(f"{name} - Class Distribution")
    if label_col not in df.columns:
        print(f"Label column '{label_col}' not found.")
        return

    counts = df[label_col].value_counts()

    if activity_map:
        counts.index = counts.index.map(lambda x: activity_map.get(x, f"Unknown ({x})"))

    print(counts)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=counts.index, y=counts.values)
    plt.title(f'Class Distribution - {name}')
    plt.ylabel('Number of Samples')
    plt.xlabel('Activity')
    plt.xticks(rotation=45, ha='right')
    plt.show()

def plot_subject_class_distribution(df, label_col, subject_col, name, activity_map=None):
    print_break(f"{name} - Subject-wise Class Distribution")

    if label_col not in df.columns or subject_col not in df.columns:
        print(f"Label or Subject column not found.")
        return

    subject_class_counts = df.groupby([subject_col, label_col]).size().reset_index(name='count')

    if activity_map:
        subject_class_counts['activity'] = subject_class_counts[label_col].map(
            lambda x: activity_map.get(x, f"Unknown ({x})")
        )
    else:
        subject_class_counts['activity'] = subject_class_counts[label_col].astype(str)

    pivot_table = subject_class_counts.pivot_table(
        index=subject_col,
        columns='activity',
        values='count',
        fill_value=0
    )

    print("\nSamples per Activity per Subject:")
    display(pivot_table)

    plt.figure(figsize=(16, 8))
    pivot_table.plot(kind='bar', stacked=False, figsize=(16, 8))
    plt.title(f'Class Distribution per Subject - {name}')
    plt.ylabel('Number of Samples')
    plt.xlabel('Subject')
    plt.legend(title='Activity', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

def plot_feature_boxplots(df, feature_cols, name, max_features=15):
    print_break(f"{name} - Feature Boxplots (First {max_features} Features)")

    plot_cols = feature_cols[:min(len(feature_cols), max_features)]

    if not plot_cols:
        print("No feature columns to plot.")
        return

    df_melted = df.melt(value_vars=plot_cols)

    plt.figure(figsize=(15, 7))
    sns.boxplot(x='variable', y='value', data=df_melted, showfliers=False)
    plt.title(f'Feature Distributions (Outliers Hidden) - {name}')
    plt.ylabel('Value')
    plt.xlabel('Features')
    plt.xticks(rotation=45, ha='right')
    plt.show()

def plot_correlation_heatmap(df, feature_cols, name):
    print_break(f"{name} - Feature Correlation Heatmap")

    if not feature_cols:
        print("No feature columns for correlation.")
        return

    corr_matrix = df[feature_cols].corr()

    plt.figure(figsize=(12, 10))
    sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt='.1f')
    plt.title(f'Feature Correlation Matrix - {name}')
    plt.show()

def plot_sensor_time_series(df, feature_cols, label_col, subject_col, name, activity_map=None, n_sensors=3, n_samples=3000):
    print_break(f"{name} - Sensor Time Series Example")

    if subject_col not in df.columns or label_col not in df.columns:
        print("Subject or Label column not found.")
        return

    try:
        subject = df[subject_col].unique()[0]
        activity = df[label_col].unique()[1]
    except IndexError:
        print("Not enough unique subjects or activities to plot.")
        return

    df_sample = df[(df[subject_col] == subject) & (df[label_col] == activity)].head(n_samples)

    if df_sample.empty:
        print(f"No data found for subject {subject} and activity {activity}.")
        return

    activity_name = activity_map.get(activity, f"Activity {activity}") if activity_map else f"Activity {activity}"
    plot_cols = feature_cols[:min(len(feature_cols), n_sensors)]

    if not plot_cols:
        print("No feature columns to plot.")
        return

    df_sample[plot_cols].plot(subplots=True, figsize=(15, 2 * n_sensors),
                             title=f'Time Series for Subject {subject} - {activity_name} ({name})')
    plt.xlabel('Time Step')
    plt.tight_layout()
    plt.show()

def plot_tsne(df, feature_cols, label_col, name, activity_map=None, n_samples=5000):
    print_break(f"{name} - t-SNE Visualization")

    df_clean = df[feature_cols + [label_col]].dropna()

    if df_clean.empty:
        print("No data left after dropping NaNs for t-SNE.")
        return

    if len(df_clean) > n_samples:
        df_sample = df_clean.sample(n=n_samples, random_state=42)
    else:
        df_sample = df_clean

    print(f"Running t-SNE on {len(df_sample)} samples...")

    tsne = TSNE(n_components=2, verbose=0, perplexity=30, n_iter=300, random_state=42)
    tsne_results = tsne.fit_transform(df_sample[feature_cols])

    df_tsne = pd.DataFrame(data=tsne_results, columns=['tsne-2d-one', 'tsne-2d-two'])
    df_tsne['label'] = df_sample[label_col].values

    if activity_map:
        df_tsne['activity'] = df_tsne['label'].map(lambda x: activity_map.get(x, f"Unknown ({x})"))
    else:
        df_tsne['activity'] = df_tsne['label'].astype(str)

    plt.figure(figsize=(14, 10))
    sns.scatterplot(
        x="tsne-2d-one", y="tsne-2d-two",
        hue="activity",
        palette=sns.color_palette("hsv", len(df_tsne['activity'].unique())),
        data=df_tsne,
        legend="full",
        alpha=0.7
    )
    plt.title(f't-SNE Projection - {name}')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()

def load_mhealth_data_raw(dataset_dir):
    subject_files = sorted([
        os.path.join(dataset_dir, f) for f in os.listdir(dataset_dir)
        if f.startswith("mHealth_subject") and f.endswith(".log")
    ])
    if not subject_files:
        print("MHEALTH files not found.")
        return None, None

    all_dfs = []
    for i, file_path in enumerate(subject_files):
        try:
            df = pd.read_csv(file_path, sep='\s+', header=None, engine='python', dtype=np.float32)
            if df.shape[1] < 24:
                continue
            df['subject'] = i + 1
            all_dfs.append(df)
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
            continue

    if not all_dfs:
        return None, None

    combined_df = pd.concat(all_dfs, ignore_index=True)
    combined_df.columns = MHEALTH_COLS + ['subject']

    feature_cols = [col for col in MHEALTH_COLS if col != 'label']
    return combined_df, feature_cols, 'label', 'subject', MHEALTH_ACTIVITY_MAPPING

def load_pamap2_data_raw(dataset_dir):
    file_paths = sorted(glob.glob(os.path.join(dataset_dir, 'Protocol', 'subject*.dat')))
    optional_path = os.path.join(dataset_dir, 'Optional')
    if os.path.exists(optional_path):
        file_paths += sorted(glob.glob(os.path.join(optional_path, 'subject*.dat')))

    if not file_paths:
        print("PAMAP2 files not found.")
        return None, None

    all_dfs = []
    for file_path in file_paths:
        subject_id_str = os.path.basename(file_path).replace('subject', '').split('.')[0]
        try:
            subject_id = int(subject_id_str)
            df = pd.read_csv(file_path, sep='\s+', header=None, na_values='NaN')
            df['subject'] = subject_id
            all_dfs.append(df)
        except Exception as e:
            print(f"Error reading {file_path}: {e}")
            continue

    if not all_dfs:
        return None, None

    combined_df = pd.concat(all_dfs, ignore_index=True)

    if combined_df.shape[1] < max([c for c in PAMAP2_COLS if isinstance(c, int)], default=0) + 1:
         combined_df.columns = PAMAP2_COLS + ['subject']
    else:
        print(f"PAMAP2: Column count mismatch. Expected {len(PAMAP2_COLS)}, got {combined_df.shape[1]}")
        temp_cols = PAMAP2_COLS + ['subject']
        combined_df = combined_df.iloc[:, :len(temp_cols)]
        combined_df.columns = temp_cols

    return combined_df, PAMAP2_FEATURE_COLS, 'label', 'subject', PAMAP2_ACTIVITY_MAPPING

def read_txt_matrix(file_path):
    try:
        return np.loadtxt(file_path, dtype=np.float32)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        raise

def load_uci_data_raw(dataset_dir):
    UCI_CHANNELS_PREFIX = [
        "total_acc_x_", "total_acc_y_", "total_acc_z_",
        "body_acc_x_", "body_acc_y_", "body_acc_z_",
        "body_gyro_x_", "body_gyro_y_", "body_gyro_z_",
    ]

    all_data = {'train': {}, 'test': {}}

    for split in ['train', 'test']:
        try:
            data_root = os.path.join(dataset_dir)
            signal_root = os.path.join(data_root)

            channels = []
            for prefix in UCI_CHANNELS_PREFIX:
                file_path = os.path.join(signal_root, f"{prefix}{split}.txt")
                if not os.path.exists(file_path):
                    raise FileNotFoundError(f"File not found: {file_path}")
                channels.append(read_txt_matrix(file_path))

            all_data[split]['X'] = np.stack(channels, axis=1)

            y_path = os.path.join(data_root, f"y_{split}.txt")
            all_data[split]['y'] = read_txt_matrix(y_path).astype(int) - 1

            subject_path = os.path.join(data_root, f"subject_{split}.txt")
            all_data[split]['subject'] = read_txt_matrix(subject_path).astype(int)

        except FileNotFoundError as e:
            print(f"UCI {split} files not found: {e}")
            return None, None
        except Exception as e:
            print(f"Error loading UCI {split} data: {e}")
            return None, None

    X_combined = np.vstack((all_data['train']['X'], all_data['test']['X']))
    y_combined = np.hstack((all_data['train']['y'], all_data['test']['y']))
    subjects_combined = np.hstack((all_data['train']['subject'], all_data['test']['subject']))

    n_samples, n_channels, n_timesteps = X_combined.shape

    X_reshaped = X_combined.transpose(0, 2, 1).reshape(-1, n_channels)

    df = pd.DataFrame(X_reshaped, columns=UCI_CHANNELS_PREFIX)

    df['label'] = np.repeat(y_combined, n_timesteps)
    df['subject'] = np.repeat(subjects_combined, n_timesteps)
    df['timestamp'] = np.tile(np.arange(n_timesteps), n_samples)

    return df, UCI_CHANNELS_PREFIX, 'label', 'subject', UCI_ACTIVITY_MAPPING

def load_wisdm_data_raw(dataset_path):
    possible_files = ['WISDM_ar_v1.1_raw.txt', 'actitracker_raw.txt']
    file_to_load = None
    for f in possible_files:
        full_path = os.path.join(dataset_path, f)
        if os.path.exists(full_path):
            file_to_load = full_path
            break

    if not file_to_load:
        print(f"WISDM raw data file not found in {dataset_path}")
        return None, None

    print(f"Loading WISDM data from: {file_to_load}")

    all_data = []
    try:
        with open(file_to_load, 'r') as f:
            lines = f.readlines()

        for line in lines:
            line = line.strip().rstrip(';').rstrip(',')
            if not line:
                continue

            parts = line.split(',')
            if len(parts) != 6:
                continue

            try:
                user = parts[0].strip()
                activity = parts[1].strip()
                timestamp = int(parts[2].strip())
                x = float(parts[3].strip())
                y = float(parts[4].strip())
                z = float(parts[5].strip())
                all_data.append([user, activity, timestamp, x, y, z])
            except (ValueError, IndexError):
                continue

    except Exception as e:
        print(f"Error reading WISDM file: {e}")
        return None, None

    if not all_data:
        print("No valid data parsed from WISDM file.")
        return None, None

    df = pd.DataFrame(all_data, columns=['subject', 'label', 'timestamp', 'x', 'y', 'z'])
    df['x'] = pd.to_numeric(df['x'], errors='coerce')
    df['y'] = pd.to_numeric(df['y'], errors='coerce')
    df['z'] = pd.to_numeric(df['z'], errors='coerce')

    feature_cols = ['x', 'y', 'z']

    le = LabelEncoder()
    df['label'] = le.fit_transform(df['label'])

    activity_map = {i: label for i, label in enumerate(le.classes_)}

    return df, feature_cols, 'label', 'subject', activity_map

def load_opportunity_local_data_raw(data_path):
    print("Loading Opportunity dataset from local files...")
    try:
        dat_files = glob.glob(os.path.join(data_path, "**", "*.dat"), recursive=True)
        if not dat_files:
            print(f"No .dat files found in {data_path}")
            return None, None

        print(f"Found {len(dat_files)} .dat files")

        all_dfs = []
        for file_path in dat_files:
            subject_id = os.path.basename(file_path).split('.')[0]
            try:
                data = pd.read_csv(file_path, sep=r'\s+', header=None, na_values='NaN')

                if data.shape[1] < 250:
                    print(f"Skipping {file_path}: Not enough columns ({data.shape[1]})")
                    continue

                data['subject'] = subject_id
                all_dfs.append(data)
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
                continue

        if not all_dfs:
            print("No valid data files loaded.")
            return None, None

        combined_df = pd.concat(all_dfs, ignore_index=True)

        feature_df = combined_df.iloc[:, OPPORTUNITY_COL_IDX]
        feature_df.columns = OPPORTUNITY_FEATURE_COLS

        label_df = combined_df.iloc[:, [OPPORTUNITY_LABEL_COL]]
        label_df.columns = ['label']

        subject_df = combined_df[['subject']]

        final_df = pd.concat([feature_df, label_df, subject_df], axis=1)

        return final_df, OPPORTUNITY_FEATURE_COLS, 'label', 'subject', OPPORTUNITY_ACTIVITY_MAPPING

    except Exception as e:
        print(f"Error loading Opportunity dataset: {e}")
        return None, None

def load_usc_had_data_raw(dataset_path):
    print("Loading USC-HAD dataset...")

    all_data = []

    mat_files = glob.glob(os.path.join(dataset_path, "**", "*.mat"), recursive=True)
    if not mat_files:
        print(f"No .mat files found in {dataset_path}")
        return None, None

    for name in mat_files:
        try:
            mat = scipy.io.loadmat(name)
            subject_id = int(mat['subject'][0][0])
            act_num = int(mat['activity_number'][0][0])
            signal = np.array(mat['sensor_readings'], dtype=np.float32)

            df = pd.DataFrame(signal, columns=['acc_x', 'acc_y', 'acc_z', 'gyro_x', 'gyro_y', 'gyro_z'])
            df['label'] = act_num
            df['subject'] = subject_id
            all_data.append(df)

        except Exception as e:
            print(f"Error loading {name}: {e}")

    if not all_data:
        print("Error: No USC-HAD data loaded")
        return None, None

    combined_df = pd.concat(all_data, ignore_index=True)
    feature_cols = ['acc_x', 'acc_y', 'acc_z', 'gyro_x', 'gyro_y', 'gyro_z']

    return combined_df, feature_cols, 'label', 'subject', USC_ACTIVITY_MAPPING

def run_eda(load_func, path, name):
    print_break(f"STARTING EDA FOR: {name}")

    try:
        result = load_func(path)
        if result is None or result[0] is None:
            print(f"Failed to load data for {name}.")
            return

        df, feature_cols, label_col, subject_col, activity_map = result

        print(f"Successfully loaded {name} data.")
        print(f"DataFrame shape: {df.shape}")

        print_break(f"{name} - Data Head")
        display(df.head())

        print_break(f"{name} - DataFrame Info")
        df.info()

        print_break(f"{name} - Descriptive Statistics")
        display(df[feature_cols].describe())

        print_file_info(path, name)

        plot_missing_values(df, name)

        plot_class_distribution(df, label_col, name, activity_map)

        plot_subject_class_distribution(df, label_col, subject_col, name, activity_map)

        # plot_feature_boxplots(df, feature_cols, name)

        # plot_sensor_time_series(df, feature_cols, label_col, subject_col, name, activity_map)

    except Exception as e:
        print(f"An error occurred during {name} EDA: {e}")

    print_break(f"COMPLETED EDA FOR: {name}")

def main():
    datasets_to_run = {
        "MHEALTH": (load_mhealth_data_raw, MHEALTH_PATH),
        "PAMAP2": (load_pamap2_data_raw, PAMAP2_PATH),
        "UCI-HAR": (load_uci_data_raw, UCI_PATH),
        "WISDM": (load_wisdm_data_raw, WISDM_PATH),
        "OPPORTUNITY": (load_opportunity_local_data_raw, OPPORTUNITY_PATH),
        "USC-HAD": (load_usc_had_data_raw, USC_PATH),
    }

    for name, (load_func, path) in datasets_to_run.items():
        if not os.path.exists(path):
            print_break(f"SKIPPING {name}: Path not found at {path}")
            continue

        run_eda(load_func, path, name)


if __name__ == "__main__":
    main()
```
