```python
import os
import time
import math
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.amp import autocast, GradScaler
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from IPython.display import display
from ptflops import get_model_complexity_info

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True

def compute_model_stats(model, input_shape=(1,9,128)):
    macs, params = get_model_complexity_info(model, (input_shape[1], input_shape[2]), as_strings=False, print_per_layer_stat=False, verbose=False)
    return params/1e6, macs/1e6

def measure_inference_time(model, sz=(1,9,128), dev='cuda', runs=100, warm=10):
    model.eval()
    x = torch.randn(sz, device=dev)
    with torch.inference_mode():
        for _ in range(warm): _ = model(x)
    torch.cuda.synchronize() if dev=='cuda' else None
    t0 = time.time()
    with torch.inference_mode():
        for _ in range(runs): _ = model(x)
    torch.cuda.synchronize() if dev=='cuda' else None
    return (time.time()-t0)/runs*1000

def plot_confusion_matrix(cm, labels, path='cm.png'):
    cmn = cm.astype('float')/cm.sum(axis=1)[:,None]
    df = pd.DataFrame(cmn, index=labels, columns=labels)
    plt.figure(figsize=(6,6))
    sns.heatmap(df, annot=True, fmt='.2f', cmap='Blues', cbar=True)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.tight_layout()
    # plt.savefig(path, dpi=300)
    plt.show()
    plt.close()

class HARDataset(Dataset):
    def __init__(self, X, y, norm=True):
        self.X = X.astype(np.float32)
        self.y = y.astype(np.int64)
        if norm:
            self.mean = self.X.mean(axis=(0,2), keepdims=True)
            self.std = self.X.std(axis=(0,2), keepdims=True) + 1e-8
            self.X = (self.X - self.mean)/self.std
        else:
            self.mean, self.std = 0, 1
    def set_stats(self, mean, std):
        self.mean, self.std = mean, std
        self.X = (self.X - self.mean)/self.std
    def __len__(self): return len(self.y)
    def __getitem__(self, i): return torch.from_numpy(self.X[i]), int(self.y[i])

def read_txt_matrix(p): return np.loadtxt(p, dtype=np.float32)

def load_full_uci_har_data(root):
    pref = ["total_acc_x_","total_acc_y_","total_acc_z_","body_acc_x_","body_acc_y_","body_acc_z_","body_gyro_x_","body_gyro_y_","body_gyro_z_"]
    def load(s):
        ch = [read_txt_matrix(os.path.join(root,f"{p}{s}.txt")) for p in pref]
        X = np.stack(ch, axis=1)
        y = read_txt_matrix(os.path.join(root,f"y_{s}.txt")).astype(int)-1
        sub = read_txt_matrix(os.path.join(root,f"subject_{s}.txt")).astype(int)
        return X,y,sub
    Xtr,ytr,str = load("train")
    Xte,yte,ste = load("test")
    X = np.concatenate((Xtr,Xte))
    y = np.concatenate((ytr,yte))
    sub = np.concatenate((str,ste))
    idx = np.argsort(sub)
    X,y,sub = X[idx],y[idx],sub[idx]
    names = ['Walking','Walking Upstairs','Walking Downstairs','Sitting','Standing','Laying']
    return X,y,sub,names

def fft_filter(x, cut, fs, typ='low'):
    B,C,T = x.shape
    f = torch.fft.fftfreq(T, d=1/fs).to(x.device)
    X = torch.fft.fft(x, dim=-1)
    mask = (torch.abs(f)<=cut) if typ=='low' else (torch.abs(f)>cut)
    mask = mask.view(1,1,-1).expand(B,C,-1)
    return torch.fft.ifft(X*mask, dim=-1).real

def compute_gravity_gyro_consistency(acc, gyro, grav, fs, eps=1e-6):
    dt = 1.0/fs
    B,C,T = acc.shape
    n = min(acc.shape[1]//3, gyro.shape[1]//3)
    loss = 0.0
    ux = torch.tensor([1.,0.,0.], device=acc.device)
    uy = torch.tensor([0.,1.,0.], device=acc.device)
    uz = torch.tensor([0.,0.,1.], device=acc.device)
    for i in range(n):
        g = grav[:,i*3:(i+1)*3,:]
        w = gyro[:,i*3:(i+1)*3,:]
        gn = F.normalize(g, dim=1, eps=eps)
        l = 0.0
        for t in range(1,T):
            gp = gn[:,:,t-1]
            gc = gn[:,:,t]
            ang = w[:,:,t-1]*dt
            pred = gp.clone()
            ax = torch.cross(gp, ux.expand_as(gp), dim=1)
            ay = torch.cross(gp, uy.expand_as(gp), dim=1)
            az = torch.cross(gp, uz.expand_as(gp), dim=1)
            # *** FIX: Changed ang[:,:,0:1] to ang[:,0:1] (and for 1:2, 2:3) ***
            pred = pred + ang[:,0:1]*ax + ang[:,1:2]*ay + ang[:,2:3]*az
            pred = F.normalize(pred, dim=1, eps=eps)
            l += F.mse_loss(pred, gc)
        loss += l/(T-1)
    return loss/n

def physics_guided_loss(x, gscale, fs=50, hp=1.5, lp=0.5, eps=1e-6):
    acc = x[:,[3,4,5],:]
    gyro = x[:,[6,7,8],:]
    low = fft_filter(acc, lp, fs, 'low')
    Lg = compute_gravity_gyro_consistency(acc, gyro, low, fs)
    ah = fft_filter(acc, hp, fs, 'high')
    gh = fft_filter(gyro, hp, fs, 'high')
    aa = (ah**2).sum(dim=1)
    ga = (gh**2).sum(dim=1)
    an = (aa-aa.mean(-1,keepdim=True))/(aa.std(-1,keepdim=True)+eps)
    gn = (ga-ga.mean(-1,keepdim=True))/(ga.std(-1,keepdim=True)+eps)
    Lag = F.mse_loss(an, gn)
    at = F.normalize(ah.flatten(1), dim=1)
    gt = F.normalize(gh.flatten(1), dim=1)
    corr = (at*gt).sum(1).mean()
    Lagc = 1.0-corr
    Lag = Lag + 2*Lagc
    aj = torch.diff(acc, dim=-1)
    gj = torch.diff(gyro, dim=-1)
    Lj = (aj**2).mean() + (gj**2).mean()
    return Lg, Lag, Lj

class ODEF(nn.Module):
    def forward_with_grad(self, z, t, grad_outputs):
        bs = z.shape[0]
        out = self.forward(z, t)
        a = grad_outputs
        adfdz, adfdt, *adfdp = torch.autograd.grad((out,), (z, t)+tuple(self.parameters()), grad_outputs=(a,), allow_unused=True, retain_graph=True)
        if adfdp: adfdp = torch.cat([p.flatten() for p in adfdp]).unsqueeze(0).expand(bs,-1)/bs
        if adfdt is not None: adfdt = adfdt.expand(bs,1)/bs
        return out, adfdz, adfdt, adfdp
    def flatten_parameters(self):
        return torch.cat([p.flatten() for p in self.parameters()])

class ODEAdjoint(torch.autograd.Function):
    @staticmethod
    def forward(ctx, z0, t, flat_params, func):
        bs, *z_shape = z0.size()
        tl = t.size(0)
        with torch.no_grad():
            z = torch.zeros(tl, bs, *z_shape, device=z0.device)
            z[0] = z0
            for i in range(tl-1):
                z0 = ode_solve(z0, t[i], t[i+1], func)
                z[i+1] = z0
        ctx.func = func
        ctx.save_for_backward(t, z.clone(), flat_params)
        return z
    @staticmethod
    def backward(ctx, dLdz):
        func = ctx.func
        t, z, flat_params = ctx.saved_tensors
        tl, bs, *z_shape = z.size()
        nd = np.prod(z_shape)
        nprm = flat_params.size(0)
        def aug_dyn(aug, ti):
            zi = aug[:,:nd].view(bs,*z_shape)
            a = aug[:,nd:2*nd].view(bs,*z_shape)
            ti = ti.float()
            with torch.enable_grad():
                ti = ti.detach().requires_grad_(True)
                zi = zi.detach().requires_grad_(True)
                f, dfdz, dfdt, dfdp = func.forward_with_grad(zi, ti, a)
                dfdz = dfdz.to(zi) if dfdz is not None else torch.zeros_like(zi)
                dfdp = dfdp.to(zi) if dfdp is not None else torch.zeros(bs, nprm, device=zi.device)
                dfdt = dfdt.to(zi) if dfdt is not None else torch.zeros(bs,1, device=zi.device)
            return torch.cat((f.view(bs,nd), -dfdz.view(bs,nd), -dfdp.view(bs,nprm), -dfdt.view(bs,1)), dim=1)
        dLdz = dLdz.view(tl, bs, nd).float()
        with torch.no_grad():
            adj_z = torch.zeros(bs, nd, device=dLdz.device)
            adj_p = torch.zeros(bs, nprm, device=dLdz.device)
            adj_t = torch.zeros(tl, bs, 1, device=dLdz.device)
            for i in range(tl-1,0,-1):
                zi = z[i].float()
                ti = t[i].float()
                fi = func(zi, ti).view(bs,nd).float()
                dLdzi = dLdz[i].float()
                dLdti = torch.bmm(dLdzi.unsqueeze(-1).transpose(1,2), fi.unsqueeze(-1))[:,0]
                adj_z += dLdzi
                adj_t[i] = adj_t[i] - dLdti
                aug0 = torch.cat((zi.view(bs,nd), adj_z, torch.zeros(bs,nprm,device=z.device), adj_t[i]), dim=-1)
                aug1 = ode_solve(aug0, ti, t[i-1], aug_dyn)
                adj_z[:] = aug1[:,nd:2*nd]
                adj_p[:] += aug1[:,2*nd:2*nd+nprm]
                adj_t[i-1] = aug1[:,2*nd+nprm:]
            dLdz0 = dLdz[0].float()
            dLdt0 = torch.bmm(dLdz0.unsqueeze(-1).transpose(1,2), fi.unsqueeze(-1))[:,0]
            adj_z += dLdz0
            adj_t[0] = adj_t[0] - dLdt0
        return adj_z.view(bs,*z_shape), adj_t, adj_p, None

def ode_solve(z0, t0, t1, f):
    hmax = 0.05
    steps = math.ceil((abs(t1-t0)/hmax).max().item())
    h = (t1-t0)/steps
    z = z0
    t = t0
    for _ in range(steps):
        z = z + h*f(z, t)
        t = t + h
    return z

class NeuralODE(nn.Module):
    def __init__(self, func):
        super().__init__()
        assert isinstance(func, ODEF)
        self.func = func
    def forward(self, z0, t=torch.tensor([0.,1.]), whole=False):
        t = t.to(z0)
        z = ODEAdjoint.apply(z0, t, self.func.flatten_parameters(), self.func)
        return z if whole else z[-1]

class NNODEF(ODEF):
    def __init__(self, dim, hid, ti=False):
        super().__init__()
        self.ti = ti
        self.l1 = nn.Linear(dim+1 if not ti else dim, hid)
        self.l2 = nn.Linear(hid, hid)
        self.l3 = nn.Linear(hid, dim)
        self.act = nn.ELU(inplace=True)
    def forward(self, x, t):
        if not self.ti:
            t = t.expand(x.shape[0],1)
            x = torch.cat((x,t), dim=-1)
        h = self.act(self.l1(x))
        h = self.act(self.l2(h))
        return self.l3(h)

class LogSigCDEFunc(ODEF):
    def __init__(self, in_dim, hid_dim):
        super().__init__()
        self.in_dim = in_dim
        self.hid_dim = hid_dim
        self.theta = nn.Parameter(torch.randn(in_dim, hid_dim))
        self.net = nn.Sequential(nn.Linear(hid_dim, hid_dim*2), nn.ELU(), nn.Linear(hid_dim*2, hid_dim))
    def forward(self, X, t=None):
        dX = X[:,1:] - X[:,:-1]
        dX = F.normalize(dX, dim=-1)
        h = F.linear(dX, self.theta)
        return self.net(h)

class NeuralCDE(nn.Module):
    def __init__(self, in_dim, hid_dim, out_dim, seq_len, **kwargs):
        super().__init__()
        self.seq_len = seq_len
        self.init = nn.Linear(in_dim, hid_dim)
        self.cdef = LogSigCDEFunc(in_dim, hid_dim)
        self.odef = NNODEF(hid_dim, hid_dim*2, ti=True)
        self.ode = NeuralODE(self.odef)
        self.norm = nn.LayerNorm(hid_dim)
        self.cls = nn.Linear(hid_dim, out_dim)
        self.gscale = nn.Parameter(torch.tensor(1.0))
    def extract_features(self, x):
        B,C,T = x.shape
        X = torch.cumsum(x, dim=2)
        z0 = self.init(X[:,:,0])
        t = torch.linspace(0,1,T, device=x.device)
        zT = self.ode(z0, t)
        return self.norm(zT)
    def forward(self, x):
        return self.cls(self.extract_features(x))

class CNNModel(nn.Module):
    def __init__(self, n_channels=9, n_classes=6, **kwargs):
        super().__init__()
        self.c1 = nn.Conv1d(n_channels,64,5,padding=2)
        self.c2 = nn.Conv1d(64,128,5,padding=2)
        self.c3 = nn.Conv1d(128,256,5,padding=2)
        self.pool = nn.AdaptiveAvgPool1d(1)
        self.cls = nn.Linear(256, n_classes)
    def extract_features(self, x):
        x = F.relu(self.c1(x))
        x = F.relu(self.c2(x))
        x = F.relu(self.c3(x))
        return self.pool(x).flatten(1)
    def forward(self, x):
        return self.cls(self.extract_features(x))

class TransformerModel(nn.Module):
    def __init__(self, n_channels=9, seq_len=128, d_model=64, n_classes=6, nhead=4, nlayers=2):
        super().__init__()
        self.proj = nn.Linear(n_channels, d_model)
        enc = nn.TransformerEncoderLayer(d_model, nhead, dim_feedforward=256, batch_first=True)
        self.trans = nn.TransformerEncoder(enc, nlayers)
        self.norm = nn.LayerNorm(d_model)
        self.cls = nn.Linear(d_model, n_classes)
    def extract_features(self, x):
        B,C,T = x.shape
        x = x.transpose(1,2)
        x = self.proj(x)
        x = self.trans(x)
        return self.norm(x.mean(1))
    def forward(self, x):
        return self.cls(self.extract_features(x))

class UCIHarODEModel(nn.Module):
    def __init__(self, n_channels=9, seq_len=128, d_model=64, n_classes=6):
        super().__init__()
        self.proj = nn.Linear(n_channels, d_model)
        self.f = NNODEF(d_model, d_model*2)
        self.ode = NeuralODE(self.f)
        self.norm = nn.LayerNorm(d_model)
        self.cls = nn.Linear(d_model, n_classes)
        self.gscale = nn.Parameter(torch.tensor(1.0))
    def extract_features(self, x):
        B,C,T = x.shape
        x = x.transpose(1,2)
        z0 = self.proj(x[:,0,:])
        t = torch.linspace(0,1,T, device=x.device)
        zT = self.ode(z0, t)
        return self.norm(zT)
    def forward(self, x):
        return self.cls(self.extract_features(x))

def evaluate(model, loader, dev):
    model.eval()
    preds, labs = [], []
    with torch.inference_mode():
        for x,y in loader:
            x,y = x.to(dev), y.to(dev)
            out = model(x)
            p = out.argmax(-1)
            preds.append(p.cpu().numpy())
            labs.append(y.cpu().numpy())
    preds = np.concatenate(preds)
    labs = np.concatenate(labs)
    return {'acc':accuracy_score(labs,preds),
            'f1':f1_score(labs,preds,average='macro'),
            'prec':precision_score(labs,preds,average='macro',zero_division=0),
            'rec':recall_score(labs,preds,average='macro',zero_division=0),
            'cm':confusion_matrix(labs,preds)}

def print_res(name, met, par, flop, inf, acts):
    print(f"\n{'='*80}\nResults: {name}\n{'='*80}")
    print(f"Acc: {met['acc']:.4f} | F1: {met['f1']:.4f} | Prec: {met['prec']:.4f} | Rec: {met['rec']:.4f}")
    print(f"Params: {par:.4f}M | FLOPs: {flop:.4f}M | Inf: {inf:.4f}ms\n{'='*80}")
    plot_confusion_matrix(met['cm'], acts, f"./cm_{name.replace(' ','_')}.png")

def train(model, trl, val, ep, lr, wd, dev, phys, args, mean, std):
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=wd)
    sched = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=ep)
    scaler = GradScaler('cuda')
    mt = torch.from_numpy(mean).to(dev)
    st = torch.from_numpy(std).to(dev)
    best, bst = 0.0, None
    for e in range(1,ep+1):
        model.train()

        epoch_loss, epoch_lg, epoch_la, epoch_lj = 0.0, 0.0, 0.0, 0.0
        num_batches = 0

        for x,y in trl:
            x,y = x.to(dev), y.to(dev)
            opt.zero_grad()

            lg_batch, la_batch, lj_batch = 0.0, 0.0, 0.0

            with autocast('cuda'):
                out = model(x)
                loss = F.cross_entropy(out, y)
                if phys and hasattr(model,'gscale'):
                    xr = x*st + mt
                    lg,la,lj = physics_guided_loss(xr, model.gscale)
                    loss += args['wg']*lg + args['wa']*la + args['wj']*lj

                    lg_batch = lg.item()
                    la_batch = la.item()
                    lj_batch = lj.item()

            scaler.scale(loss).backward()
            scaler.unscale_(opt)
            torch.nn.utils.clip_grad_norm_(model.parameters(),1.0)
            scaler.step(opt)
            scaler.update()

            epoch_loss += loss.item()
            epoch_lg += lg_batch
            epoch_la += la_batch
            epoch_lj += lj_batch
            num_batches += 1

        sched.step()
        vm = evaluate(model, val, dev)

        avg_loss = epoch_loss / num_batches
        avg_lg = epoch_lg / num_batches
        avg_la = epoch_la / num_batches
        avg_lj = epoch_lj / num_batches
        val_acc = vm['acc']

        log_msg = f"Epoch {e}/{ep} | Val Acc: {val_acc:.4f} | Total Loss: {avg_loss:.4f}"
        if phys:
            log_msg += f" | Lg: {avg_lg:.4f} | La: {avg_la:.4f} | Lj: {avg_lj:.4f}"
        print(log_msg)

        if vm['acc']>best:
            best = vm['acc']
            bst = {k:v.cpu() for k,v in model.state_dict().items()}

    if bst: model.load_state_dict({k:v.to(dev) for k,v in bst.items()})
    return model

def run_exp(name, cls, args, trl, val, tel, ep, dev, acts, mean, std, resl, phys):
    print(f"\n{'#'*80}\n# {name}\n{'#'*80}")
    margs = {k:v for k,v in args.items() if k in ['n_channels','n_classes','seq_len','d_model','nhead','nlayers','in_dim','hid_dim','out_dim']}
    m = cls(**margs).to(dev)
    m = train(m, trl, val, ep, args['lr'], args['wd'], dev, phys, args, mean, std)
    met = evaluate(m, tel, dev)
    par, flop = compute_model_stats(m)
    inf = measure_inference_time(m, dev=dev)
    print_res(name, met, par, flop, inf, acts)
    resl.append({'Exp':name,'Acc':f"{met['acc']:.4f}",'F1':f"{met['f1']:.4f}",
                 'Prec':f"{met['prec']:.4f}",'Rec':f"{met['rec']:.4f}",
                 'Par(M)':f"{par:.4f}",'FLOPs(M)':f"{flop:.4f}",'Inf(ms)':f"{inf:.4f}"})
    return m

def add_noise(X, typ, lvl, seed=None):
    if seed: np.random.seed(seed)
    Xn = X.copy()
    if lvl<=0: return Xn
    B,C,T = X.shape
    if typ=='jitter':
        for b in range(B):
            for c in range(C):
                js = lvl*50
                tn = np.random.normal(0,js,T)
                ti = np.arange(T)+tn
                ti = np.clip(ti,0,T-1)
                Xn[b,c,:] = np.interp(ti, np.arange(T), X[b,c,:]).astype(np.float32)
    elif typ=='dropout':
        mask = np.random.random(X.shape)<lvl
        Xn[mask]=0.0
    elif typ=='window_shifting':
        ws = max(1,int(T*0.3))
        ms = max(1,int(lvl*T*0.8))
        for b in range(B):
            for c in range(C):
                nw = T//ws
                for w in range(nw):
                    s = w*ws
                    e = min((w+1)*ws,T)
                    if np.random.random()<lvl:
                        sh = np.random.randint(-ms,ms+1)
                        if sh>0:
                            Xn[b,c,s:e-sh] = X[b,c,s+sh:e]
                            Xn[b,c,e-sh:e] = X[b,c,e-1]
                        elif sh<0:
                            Xn[b,c,s-sh:e] = X[b,c,s:e+sh]
                            Xn[b,c,s:s-sh] = X[b,c,s]
    return Xn

if __name__=="__main__":
    set_seed(42)
    dev = 'cuda' if torch.cuda.is_available() else 'cpu'
    root = "/content/drive/MyDrive/HAR_Dataset/UCI"
    ep = 100
    bs = 512
    nw = 1
    base = {'n_channels':9,'seq_len':128,'d_model':64,'n_classes':6,
            'lr':1e-3,'wd':1e-2,'wg':0.05,'wa':0.05,'wj':0.01}
    X,y,sub,acts = load_full_uci_har_data(root)
    tst_sub = [3,12,13,15,16,17,21,26,27]
    trm = ~np.isin(sub,tst_sub)
    tem = np.isin(sub,tst_sub)
    Xtr,ytr = X[trm],y[trm]
    Xte,yte = X[tem],y[tem]
    res = []
    cfgs = [
        {'name':'CNN','cls':CNNModel,'args':base,'phys':False},
        {'name':'Transformer','cls':TransformerModel,'args':base,'phys':False},
        {'name':'NeuralCDE+Physics','cls':NeuralCDE,'args':{**base,'in_dim':9,'hid_dim':64,'out_dim':6},'phys':True},
        {'name':'NeuralODE+Physics','cls':UCIHarODEModel,'args':base,'phys':True}
    ]
    mods = {}
    nstats = None
    for cfg in cfgs:
        kf = KFold(5,shuffle=True,random_state=42)
        tri,vali = next(kf.split(Xtr,ytr))
        trds = HARDataset(Xtr[tri],ytr[tri],True)
        vads = HARDataset(Xtr[vali],ytr[vali],False)
        teds = HARDataset(Xte,yte,False)
        vads.set_stats(trds.mean,trds.std)
        teds.set_stats(trds.mean,trds.std)
        nstats = (trds.mean,trds.std)
        trl = DataLoader(trds,bs,True,num_workers=nw,pin_memory=True)
        val = DataLoader(vads,bs,False,num_workers=nw,pin_memory=True)
        tel = DataLoader(teds,bs,False,num_workers=nw,pin_memory=True)
        m = run_exp(cfg['name'],cfg['cls'],cfg['args'],trl,val,tel,ep,dev,acts,trds.mean,trds.std,res,cfg['phys'])
        mods[cfg['name']] = m
    ntyp = ['jitter','dropout','window_shifting']
    nlev = [0.0,0.05,0.10,0.15,0.20,0.25,0.30]
    nlab = ['0%','5%','10%','15%','20%','25%','30%']
    for nt in ntyp:
        nr = []
        for lv,la in zip(nlev,nlab):
            row = {'Noise':la}
            for mn,m in mods.items():
                Xn = add_noise(Xte,nt,lv,42)
                tdn = HARDataset(Xn,yte,False)
                tdn.set_stats(*nstats)
                tln = DataLoader(tdn,bs,False,num_workers=nw,pin_memory=True)
                met = evaluate(m,tln,dev)
                row[f'{mn}_Acc'] = f"{met['acc']:.4f}"
                row[f'{mn}_F1'] = f"{met['f1']:.4f}"
            nr.append(row)
        ret = {'Noise':'Avg_Ret'}
        for mn in mods:
            acs = [float(r[f'{mn}_Acc']) for r in nr[1:]]
            base = float(nr[0][f'{mn}_Acc'])
            ret[f'{mn}_Acc'] = f"{np.mean([a/base for a in acs]):.4f}" if base>0 else 'ERR'
            ret[f'{mn}_F1'] = '-'
        nr.append(ret)
        dfn = pd.DataFrame(nr)
        display(dfn)
        dfn.to_csv(f'./{nt}_noise.csv',index=False)
    df = pd.DataFrame(res)
    display(df)
    df.to_csv('./summary.csv',index=False)
```
